```python
def compute(a,b):
  try:
    answer = ((a**2) + (b**2)) / (a - b)
    return answer
  except ZeroDivisionError:
    return "Error: Division not possible"
  
a, b = map(int,input().split())

print(compute(a,b))
```

```python
with open('data.txt','r') as file:
  txt = file.read()
  txt = txt.lower()
  
  txt = txt.replace('.',' ')
  txt = txt.replace('?',' ')
  txt = txt.replace('!',' ')
  txt = txt.replace(',',' ')
  txt = txt.replace(';',' ')
  txt = txt.replace("'"," ")
  txt = txt.replace(':',' ')
  
  txt_list = txt.split()
  
  dictionary = {}
  
  for i in txt_list:
    if i not in dictionary.keys():
      dictionary[i] = 1
    else:
      dictionary[i] += 1
  
  maximum_string = dict(sorted(dictionary.items(),key=lambda x:x[1], reverse=True))
  
  print(next(iter(maximum_string.keys())))
```

```python
number = int(input())

string = str(number)
reverse_number = int(string[::-1])

if number == reverse_number:
  print("Mirror Number")
else:
  print("Not Mirror Number")
```
