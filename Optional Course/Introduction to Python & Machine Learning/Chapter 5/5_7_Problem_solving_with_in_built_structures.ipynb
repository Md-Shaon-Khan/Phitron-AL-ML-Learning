```python
string = """Descriptive text is a type of writing that vividly depicts a specific person, place, or thing, aiming to help the reader visualize the subject using sensory details. It focuses on specific characteristics, features, and qualities—such as appearance, color, or habit—often using adjectives and simple present tense."""

words = string.split()

print(words)
```

```python
count = {}

for word in words:
  count[word] = count.get(word,0) + 1 

print(count) 
```

```python
for k,v in count.items():
  print(k,v)
```
