## Tuples

```python
tup = (10,20,30,40)

type(tup),tup
```

```python
float_tup = (10.5,11.2,14,15.9)

type(float_tup), type(float_tup[1])
```

```python
tup1 = (10)

type(tup1)
```

```python
tup1 = (10,)

type(tup1)
```

```python
tup1 = tuple([10])

type(tup1)
```

```python
lis = [10,20,40,50]

type(lis)
```

```python
lis = tuple(lis)

type(lis)
```

```python
lis = list(lis)

type(lis)
```

```python
tup = (10, 20, 30, 40.5, "Shaon", True)

print(tup[3])
```

```python
new_tup = tup[0:3]

new_tup
```

```python
type(new_tup)
```

```python
lst = [10, 20, 30, 40, 50]

lst.append(80)

lst
```

```python
lst = tuple(lst)
```

```python
lst.append(90)

lst
```
