```python
s = "Either I win or Die!Win"

print(s.lower())           # either i win or die!win
print(s.upper())           # EITHER I WIN OR DIE!WIN
print(s.capitalize())      # Either i win or die!win
print(s.title())           # Either I Win Or Die!Win
print(s.strip())           # removes leading/trailing spaces
print(len(s))              # 23

ps = s.lower()
print("win"  in ps)         # True  — membership check
print(ps.find("win"))      # 9     — first occurrence index
print(ps.rfind("win"))     # 20    — last occurrence index
print(ps.count("win"))     # 2     — how many times
print(s.replace("win", "WIN"))  # Either I WIN or Die!Win
```

```python
s1 = {1, 2, 3, 3}
s2 = set([3, 4, 5])
s1.add(6)
s1.discard(1)

print(s1 | s2) # Union
print(s1 & s2) # Intersection
print(s1 - s2) # Difference
print(s1 ^ s2) # Symmetric Difference
print(s1.issubset(s2))
```
