```python
n = int(input())

lst = []
for i in range(n):
  name = input()
  num  = float(input())
  lst.append([name,num])

lst.sort(key=lambda x:x[1])

marks = sorted(set(x[1] for x in lst))

second_lowest = marks[1]

nme = []

for i in range(n):
  if lst[i][1] == second_lowest:
    nme.append(lst[i][0])
nme = sorted(nme)

for name in nme:
    print(name)
```

```python
n = int(input())

marks_dict = {}

for _ in range(n):
    data = input().split()
    name = data[0]
    scores = list(map(float, data[1:]))
    marks_dict[name] = scores

name = input()

lst = marks_dict[name]
avg = sum(lst) / len(lst)
print(f"{avg:.2f}")
```

```python
n = int(input())

lst = []

for _ in range(n):
  command = input().split()
  
  op = command[0]
  
  if op == "insert":
    index = int(command[1])
    value = int(command[2])
    lst.insert(index,value)
    
  elif op == "print":
        print(lst)
  elif op == "remove":
        value = int(command[1])
        lst.remove(value)
  elif op == "append":
        value = int(command[1])
        lst.append(value)
  elif op == "sort":
        lst.sort()
  elif op == "pop":
        lst.pop()
  elif op == "reverse":
        lst.reverse()
        
```

```python
def mutate_string(string, position, character):
    string = list(string)
    string[position] = character
    string = ''.join(string)
    return string

if __name__ == '__main__':
    s = input()
    i, c = input().split()
    s_new = mutate_string(s, int(i), c)
    print(s_new)
```

```python
s = input()


has_alnum = False  # checks a-z, A-Z, 0-9
has_alpha = False  # checks a-z, A-Z
has_digit = False  # checks 0-9
has_lower = False  # checks a-z
has_upper = False  # checks A-Z


for c in s:
    if c.isalnum():  # alphanumeric: a-z, A-Z, 0-9
        has_alnum = True
    if c.isalpha():  # alphabetical: a-z, A-Z
        has_alpha = True
    if c.isdigit():  # digits: 0-9
        has_digit = True
    if c.islower():  # lowercase letters: a-z
        has_lower = True
    if c.isupper():  # uppercase letters: A-Z
        has_upper = True

print(has_alnum)
print(has_alpha)
print(has_digit)
print(has_lower)
print(has_upper)
```
