```python
n = int(input())

lst = []
for i in range(n):
  name = input()
  num  = float(input())
  lst.append([name,num])

lst.sort(key=lambda x:x[1])

marks = sorted(set(x[1] for x in lst))

second_lowest = marks[1]

nme = []

for i in range(n):
  if lst[i][1] == second_lowest:
    nme.append(lst[i][0])
nme = sorted(nme)

for name in nme:
    print(name)
```

```python
s = input()


has_alnum = False  # checks a-z, A-Z, 0-9
has_alpha = False  # checks a-z, A-Z
has_digit = False  # checks 0-9
has_lower = False  # checks a-z
has_upper = False  # checks A-Z


for c in s:
    if c.isalnum():  # alphanumeric: a-z, A-Z, 0-9
        has_alnum = True
    if c.isalpha():  # alphabetical: a-z, A-Z
        has_alpha = True
    if c.isdigit():  # digits: 0-9
        has_digit = True
    if c.islower():  # lowercase letters: a-z
        has_lower = True
    if c.isupper():  # uppercase letters: A-Z
        has_upper = True

print(has_alnum)
print(has_alpha)
print(has_digit)
print(has_lower)
print(has_upper)
```

```python
def minion_game(strg):
    n = len(strg)
    man1 = 0
    man2 = 0
    for i in range(n):
      if (strg[i] == 'A') or (strg[i] == 'E') or (strg[i] == 'I') or (strg[i] == 'O') or (strg[i] == 'U'):
        man1 += (n-i)
      else:
        man2 += (n-i)
      
    if man1 > man2:
        print(f"Kevin {man1}")
    elif man2 > man1:
        print(f"Stuart {man2}")
    else:
        print("Draw")

if __name__ == '__main__':
    s = input()
    minion_game(s)
```

```python
def merge_the_tools(string, k):
    n = len(string)
    
    for i in range(0, n, k): 
        chunk = string[i : i + k]       
        seen = "" 
        for char in chunk:     
            if char not in seen:
                seen += char        
        print(seen)

if __name__ == '__main__':
    string, k = input(), int(input())
    merge_the_tools(string, k)
```

```python
from datetime import datetime

def time_delta(t1, t2):
   
    # %a = Day, %d = date, %b = Month, %Y = Year, %H:%M:%S = Time, %z = Timezone
    time_format = '%a %d %b %Y %H:%M:%S %z'
    

    time1 = datetime.strptime(t1, time_format)
    time2 = datetime.strptime(t2, time_format)
    
    diff = int(abs((time1 - time2).total_seconds()))
    
    return str(diff)

if __name__ == '__main__':

    t = int(input())

    for _ in range(t):
        t1 = input()
        t2 = input()

        result = time_delta(t1, t2)
        print(result)
```

```python
import operator

def person_lister(f):
    def inner(people):
        people.sort(key=lambda x: int(x[2]))
        return [f(person) for person in people]
    return inner

@person_lister
def name_format(person):
    prefix = "Mr. " if person[3] == "M" else "Ms. "
    return prefix + person[0] + " " + person[1]

if __name__ == '__main__':
    n = int(input())
    people = [input().split() for _ in range(n)]
    result = name_format(people)
    for r in result:
        print(r)
```

```python
def count_substring(string, sub_string):
    count = 0

    for i in range(len(string)):
        if string[i : i + len(sub_string)] == sub_string:
            count += 1
    return count

if __name__ == '__main__':
    string = input().strip()
    sub_string = input().strip()
    
    count = count_substring(string, sub_string)
    print(count)
```

```python
def print_formatted(number):
    width = len(bin(number)) - 2
    
    for i in range(1, number + 1):
        print("{0:{w}d} {0:{w}o} {0:{w}X} {0:{w}b}".format(i, w=width))

if __name__ == '__main__':
    n = int(input())
    print_formatted(n)
```
