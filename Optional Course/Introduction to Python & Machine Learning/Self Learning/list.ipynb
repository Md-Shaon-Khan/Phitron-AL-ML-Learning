A list is an ordered, mutable (changeable) collection that can hold items of any type — numbers, strings, even other lists. Lists use square brackets [...].

Just like strings, lists support positive and negative indexing, and slicing works exactly the same way.

![alt text](<Screenshot 2026-03-13 123616.png>)

A Stack is Last In First Out (LIFO) — like a pile of plates. A Queue is First In First Out (FIFO) — like a line of people.

## INDEXING & SLICING

```python
fruits = ["apple", "banana", "cherry", "date", "elderberry"]
```

```python
fruits[0] # Accessing the first element apple
```

```python
fruits[-1] # Accessing the last element elderberry
```

```python
fruits[1:3] # Accessing a slice of the list ['banana', 'cherry']
```

```python
fruits[::-1] # Reversing the list ['elderberry', 'date', 'cherry', 'banana', 'apple']
```

```python
fruits[-3:-1] # Accessing a slice using negative indices ['cherry', 'date']
```

Lists are MUTABLE

```python
fruits[0] = "avocado"
```

```python
fruits
```

## LIST COMPREHENSION

```python
squares = [x**2 for x in range(10)]
print(squares)
```

```python
even_squares = [x**2 for x in range(10) if x % 2 == 0]
print(even_squares)
```

```python
names = ["alice", "bob", "charlie"]
upper_names = [name.upper() for name in names]
print(upper_names)
```

```python
numbers = [1, -2, 3, -4, 5, -6]
positives = [n for n in numbers if n > 0]
print(positives)
```

```python
matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
flat   = [num for row in matrix for num in row]
print(flat)
```

## Stack

```python
stack = []
stack.append(10)   # push
stack.append(20)
stack.append(30)
print(stack)
```

```python
top = stack.pop()  # pop from top 
print(top)      
print(stack)  
```

## Queue

```python
queue = []
queue.append("A")   # enqueue
queue.append("B")
queue.append("C")
first = queue.pop(0)  # dequeue from front
print(first) 
```
