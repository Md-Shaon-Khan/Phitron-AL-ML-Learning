A string is a sequence of characters enclosed in single '...', double "...", or triple '''...''' quotes. Strings are immutable — once created, they cannot be changed in place.

1. Indexing:
Each character in a string has a position (index). Python supports both positive and negative indexing.

![alt text](<Screenshot 2026-03-13 122356-1.png>)

2. Slicing:
Slicing extracts a substring using string[start:stop:step]. The start is inclusive, stop is exclusive.

![alt text](<Screenshot 2026-03-13 122403.png>)

```python
s = "Hello world!"
```

## Slicing

```python
s[0],s[-1],s[7]
```

```python
#(index 0 to 4)

s[0:5]
```

```python
#  (from 7 to end)

s[7:]
```

```python
# (from start to 4)
s[:5]
```

```python
s[::2]
```

```python
s[::-1]
```

```python
string = list(map(str,input().split(",")))

list1 = []

for i in string:
    list1.append(i.strip())

for record in list1:
  parts = record.split(" ")
  name = parts[0]
  score = parts[1]
  grade = parts[2]
  print(f"{name.title()} passed with score {score} and grade {grade}")
```
