## Vectors

Scalar vs Vector

A scalar has only magnitude (a single number). A vector has both magnitude AND direction.

Scalar examples: 5, −3, 100, 3.14

Vector notation: v = [3, 4] in 2D — v = [x, y, z] in 3D

A vector is an arrow starting from a point, going in a direction.


## Distance from Origin (Magnitude)

The length of a vector — how far the tip is from the origin (0,0).

Formula (2D):

|v| = √(x² + y²)

Formula (3D):

|v| = √(x² + y² + z²)

Formula (nD):

|v| = √(x₁² + x₂² + x₃² + … + xₙ²)

Example: v = [3, 4] → |v| = √(9 + 16) = √25 = 5

This is just the Pythagorean theorem applied to vectors!

## Euclidean Distance

Distance between two points A and B — not from origin.

Formula (2D):

d(A,B) = √((x₂−x₁)² + (y₂−y₁)²)

Formula (nD):

d(A,B) = √( Σ (aᵢ − bᵢ)² )


Distance from origin = magnitude of a single vector

Euclidean distance = magnitude of the difference between two vectors

## Derivatives & Error Metrics

What is a Function?

A function takes an input x and gives an output y. It is a rule that maps every input to exactly one output.

f(x) = x² means — give me any x, I square it and return the result.

What is a Derivative?

A derivative tells you the rate of change of a function — how much y changes when x changes by a tiny bit.

Simple words: The derivative is the slope of the curve at a point.

Formal definition:

f'(x) = lim[h→0] (f(x+h) − f(x)) / h

This just means — take a tiny step h, measure how much f changed, divide by h, then shrink h to zero.

## Geometrical Intuition of Derivative

## Derivative of a Constant

A constant never changes — so its rate of change is always zero.

f(x) = c → f'(x) = 0


## Derivative of a Straight Line

A straight line f(x) = mx + c has a constant slope m everywhere.

f(x) = mx + c → f'(x) = m

Example: Find the derivative of f(x) = 3x² + 2x + 5

Step 1: derivative of 3x² → 3 × 2x¹ = 6x

Step 2: derivative of 2x → 2

Step 3: derivative of 5 → 0

Answer: f'(x) = 6x + 2

What does it mean?

At x = 1 → slope = 6(1) + 2 = 8

At x = 0 → slope = 6(0) + 2 = 2

At x = -1 → slope = 6(-1) + 2 = -4

## Derivatives 

1. Derivative of a Cubic Functionf(x) = x³ — apply the power rule: bring power down, reduce power by 1.

f(x) = x³ → f'(x) = 3x²

Key observation — f'(x) = 3x² is always ≥ 0, meaning the cubic function never goes downhill — it only goes up (or stays flat at x=0).

2. Generalization — Power Rule for Polynomials

One rule covers all powers. This is the most used rule in calculus.

Power Rule: f(x) = xⁿ → f'(x) = n · xⁿ⁻¹

For a full polynomial:

f(x) = aₙxⁿ + aₙ₋₁xⁿ⁻¹ + … + a₁x + a₀

f'(x) = n·aₙxⁿ⁻¹ + (n-1)·aₙ₋₁xⁿ⁻² + … + a₁

3. Derivative of log(x)

f(x) = ln(x) → f'(x) = 1/x

f(x) = log(x) → f'(x) = 1 / (x · ln(x))

4. Scalar Multiplication & Sum Rule

These are the simplest rules — apply power rule term by term.

Scalar Rule: d/dx [c·f(x)] = c · f'(x)

Sum Rule: d/dx [f(x) + g(x)] = f'(x) + g'(x)

Difference Rule: d/dx [f(x) − g(x)] = f'(x) − g'(x)

5. Product Rule

When two functions are multiplied together — you cannot just differentiate each separately.

Product Rule: d/dx [f(x) · g(x)] = f'(x)·g(x) + f(x)·g'(x)

Memory trick: "first times derivative of second + second times derivative of first"

6. Chain Rule

When a function is inside another function — the most important rule in ML/deep learning.

Chain Rule: d/dx [f(g(x))] = f'(g(x)) · g'(x)

Memory trick: "derivative of outside × derivative of inside"

7. Mathematical View of Quadratic Derivative

A deep look at f(x) = ax² + bx + c — visually and mathematically.

f'(x) = 2ax + b
At vertex (minimum/maximum): f'(x) = 0 → x = −b / 2a

8. Derivative of a Scalar Function (Constant Multiple)

d/dx [k · f(x)] = k · f'(x)

The scalar k just multiplies through — it scales the derivative but does not affect where derivatives are zero.

All these curves — 2x², 0.5x², −x² — have their minimum/maximum at the same point x=0. The scalar shifts the steepness but not the location of the critical point.


## Gradient Descent 

1. Derivative as an Optimizer

The derivative tells you the slope at any point. If slope is:

Positive → function is going up → move left to reduce

Negative → function is going down → move right to reduce

Zero → you are at the minimum → stop!

This is exactly how ML finds the best weights — follow the slope downhill until you reach the bottom.

Key idea:

If f'(x) > 0 → x is too big → decrease x

If f'(x) < 0 → x is too small → increase x

If f'(x) = 0 → x is at minimum → done

2. Cost Function

A cost function (also called loss function) measures how wrong the model's predictions are. The goal of ML is to minimize this function.

The most common cost function is MSE:

J(w, b) = (1/2n) × Σ (yᵢ − ŷᵢ)²

Where:

n = number of data points

yᵢ = actual value

ŷᵢ = w·xᵢ + b = predicted value

w = weight, b = bias

We use 1/2n instead of 1/n just to make the derivative cleaner (the 2 cancels out).

Expanding the prediction:

J(w, b) = (1/2n) × Σ (yᵢ − (w·xᵢ + b))²

This is a bowl-shaped curve — it has one clear minimum. Our job is to find the w and b at the bottom of that bowl.

3. What is Gradient Descent?

Gradient descent is an optimization algorithm that finds the minimum of a function by repeatedly taking small steps in the direction of the negative gradient (downhill).

Think of it like this:

You are blindfolded on a hilly landscape. You want to reach the lowest point. You feel the slope under your feet and take a small step in the downhill direction. Repeat until flat ground.

Update rule:

x_new = x_old − α × f'(x_old)

Where α (alpha) is the learning rate — how big each step is.

4. Procedure of Gradient Descent

Step by step:

Step 1 — Initialize

   Pick starting values for w and b (usually 0 or small random)

   w = 0,  b = 0

Step 2 — Compute prediction

   ŷᵢ = w·xᵢ + b

Step 3 — Compute cost

   J(w,b) = (1/2n) Σ (yᵢ − ŷᵢ)²

Step 4 — Compute gradients (derivatives of J)

   ∂J/∂w = (1/n) Σ (ŷᵢ − yᵢ) · xᵢ

   ∂J/∂b = (1/n) Σ (ŷᵢ − yᵢ)

Step 5 — Update weights

   w = w − α × ∂J/∂w

   b = b − α × ∂J/∂b

Step 6 — Repeat steps 2→5

   Until cost J stops decreasing (converges)

5. Deriving the Gradients

Let's derive ∂J/∂w properly from scratch.
Given:

J(w,b) = (1/2n) Σ (yᵢ − (w·xᵢ + b))²

Let eᵢ = ŷᵢ − yᵢ = (w·xᵢ + b) − yᵢ (error)

Derivative with respect to w:

Apply chain rule — outer function is u², inner is the error:

∂J/∂w = (1/2n) × Σ 2·eᵢ × ∂eᵢ/∂w

Since ∂eᵢ/∂w = xᵢ:

∂J/∂w = (1/n) × Σ eᵢ · xᵢ

∂J/∂w = (1/n) × Σ (ŷᵢ − yᵢ) · xᵢ

Derivative with respect to b:

Since ∂eᵢ/∂b = 1:

∂J/∂b = (1/n) × Σ eᵢ · 1

∂J/∂b = (1/n) × Σ (ŷᵢ − yᵢ)

Final update rules:

w = w − α × (1/n) Σ (ŷᵢ − yᵢ) · xᵢ

b = b − α × (1/n) Σ (ŷᵢ − yᵢ)

6. Learning Rate (α)




Rule of thumb values to try: 0.1, 0.01, 0.001, 0.0001
Mathematically — convergence condition:
For gradient descent to converge on MSE, the learning rate should satisfy:

0 < α < 2 / λ_max

Where λ_max is the largest eigenvalue of the data matrix XᵀX. In practice just tune it manually.

7. Numerical Example

Data: 3 points — (1,2), (2,4), (3,5)

Start: w=0, b=0, α=0.1

Iteration 1:

Predictions:

  ŷ₁ = 0×1 + 0 = 0

  ŷ₂ = 0×2 + 0 = 0

  ŷ₃ = 0×3 + 0 = 0

Errors (ŷ - y):

  e₁ = 0 − 2 = −2

  e₂ = 0 − 4 = −4

  e₃ = 0 − 5 = −5

Gradients:

  ∂J/∂w = (1/3)(e₁·x₁ + e₂·x₂ + e₃·x₃)

         = (1/3)(−2×1 + −4×2 + −5×3)

         = (1/3)(−2 − 8 − 15)

         = −25/3 = −8.33

  ∂J/∂b = (1/3)(−2 + −4 + −5)

         = −11/3 = −3.67

Updates:

  w = 0 − 0.1 × (−8.33) = 0.833

  b = 0 − 0.1 × (−3.67) = 0.367

Iteration 2:

Predictions:

  ŷ₁ = 0.833×1 + 0.367 = 1.200

  ŷ₂ = 0.833×2 + 0.367 = 2.033

  ŷ₃ = 0.833×3 + 0.367 = 2.866

Cost J = (1/6)[(2−1.2)² + (4−2.033)² + (5−2.866)²]

       = (1/6)[0.64 + 3.87 + 4.55]

       = 1.51   ← already much lower than before!
       

Keep iterating until J stops changing...

8. Types of Gradient Descent

There are 3 variants — they differ in how much data they use per update step.

Batch Gradient Descent:

Uses ALL n data points per update
∂J/∂w = (1/n) Σᵢ (ŷᵢ − yᵢ) · xᵢ

Slow for large data — but very stable

Stochastic Gradient Descent (SGD):

Uses only 1 random point per update
∂J/∂w = (ŷᵢ − yᵢ) · xᵢ

Very fast — but noisy / jumpy

Mini-batch Gradient Descent (most common):

Uses a batch of k points (e.g. k=32) per update
∂J/∂w = (1/k) Σᵢ∈batch (ŷᵢ − yᵢ) · xᵢ

Balance between speed and stability

9. Gradient Descent with Multiple Variables

When you have n features, the model becomes:

ŷ = w₁x₁ + w₂x₂ + w₃x₃ + … + wₙxₙ + b

The cost function is the same MSE — but now it is a function of n+1 parameters.

Gradient for each weight:

∂J/∂wⱼ = (1/n) Σᵢ (ŷᵢ − yᵢ) · xᵢⱼ

Where xᵢⱼ = value of feature j for data point i.

Update rule for each weight:

wⱼ = wⱼ − α × ∂J/∂wⱼ   for j = 1, 2, …, n

b  = b  − α × ∂J/∂b

In matrix form (compact and efficient):

ŷ = X·w + b            (n×1 vector of predictions)

error = ŷ − y          (n×1 error vector)

∇w = (1/n) Xᵀ · error  (gradient vector)

w = w − α × ∇w         (update all weights at once)

Where:

X is the n × m feature matrix

w is the m × 1 weight vector

Xᵀ is the transpose of X

This is why matrix operations and vectors matter so much in ML — they let you update ALL weights in one line instead of a loop.

## Statistics

What is Statistics?

Statistics is the science of collecting, organizing, analyzing, interpreting, and presenting data. It gives us tools to make sense of raw numbers and draw meaningful conclusions.

Without statistics, data is just noise. Statistics turns noise into knowledge.

## Types of Statistics

There are two main branches:

1. Descriptive Statistics

Summarizes and describes data you already have. It does not make predictions — just tells you what the data looks like.

Examples: mean, median, mode, variance, standard deviation, histograms, charts

2. Inferential Statistics

Uses a sample to make conclusions about a larger population. It makes predictions and tests hypotheses.

Examples: hypothesis testing, confidence intervals, regression, p-values

Descriptive          →    "The average score is 75"

Inferential          →    "Based on 100 students, we estimate
                           the national average is 73 ± 2"

## Population vs Sample

This is the most fundamental concept in statistics.

Population — the entire group you want to study.

All students in Bangladesh, all patients with diabetes, all transactions in a bank

Sample — a subset of the population you actually measure.

500 randomly selected students, 200 patients, 1000 transactions

Why sample?

Measuring the entire population is usually impossible, too expensive, or too slow. A good sample represents the population well.

Sample must be random and representative — otherwise results are biased.

## Mean (Average)

The sum of all values divided by the count. The central balancing point of the data.

Population mean:

μ = (1/N) × Σ xᵢ

Sample mean:

x̄ = (1/n) × Σ xᵢ

Example:

Data: 4, 8, 6, 10, 2
x̄ = (4+8+6+10+2) / 5 = 30 / 5 = 6

Weakness: Very sensitive to outliers.

Data: 4, 8, 6, 10, 200

x̄ = 228/5 = 45.6 — pulled far up by 200, not representative!

## Median

The middle value when data is sorted. Splits data into two equal halves.

If n is odd:

Median = middle value

If n is even:

Median = (n/2-th value + (n/2+1)-th value) / 2

Example (odd):

Data: 3, 7, 2, 9, 5 → sorted: 2, 3, 5, 7, 9

Median = 5 (middle value)

Example (even):

Data: 3, 7, 2, 9, 5, 11 → sorted: 2, 3, 5, 7, 9, 11

Median = (5 + 7) / 2 = 6

Strength: Not affected by outliers — better than mean for skewed data.

Data: 4, 8, 6, 10, 200

Sorted: 4, 6, 8, 10, 200 → Median = 8 (still sensible!)

## Mode

The most frequently occurring value in the dataset.

Example 1 — one mode (unimodal):

Data: 2, 3, 3, 4, 5, 3, 7

Mode = 3 (appears 3 times)

Example 2 — two modes (bimodal):

Data: 2, 2, 3, 4, 4, 5

Mode = 2 and 4

Example 3 — no mode:

Data: 1, 2, 3, 4, 5 (all appear once)
No mode

When to use:

Best for categorical data. Example: most common eye color, most sold product.



Relationship in a normal distribution:

Mean = Median = Mode (all equal — perfectly symmetric)

In a right-skewed distribution:

Mode < Median < Mean (mean pulled right by large values)

## Variance

Measures how spread out the data is from the mean. Large variance = data is all over the place. Small variance = data is clustered near the mean.

Population variance:

σ² = (1/N) × Σ (xᵢ − μ)²

Sample variance (use n-1 to correct for bias — called Bessel's correction):

s² = (1/(n-1)) × Σ (xᵢ − x̄)²

Why n-1 for sample?

Using n underestimates the true population variance. Using n-1 corrects this bias. This is called Bessel's correction.

Example:

Data: 2, 4, 4, 4, 5, 5, 7, 9

Mean: x̄ = 40/8 = 5

Deviations from mean:

(2−5)² = 9

(4−5)² = 1

(4−5)² = 1

(4−5)² = 1

(5−5)² = 0

(5−5)² = 0

(7−5)² = 4

(9−5)² = 16

Sum = 32

Population variance: σ² = 32/8 = 4

Sample variance: s² = 32/7 = 4.57

Weakness: Unit is squared — hard to interpret directly (e.g. meters² instead of meters).

## Standard Deviation

The square root of variance — brings the unit back to the original scale.

Population standard deviation:

σ = √[ (1/N) × Σ (xᵢ − μ)² ]

Sample standard deviation:

s = √[ (1/(n-1)) × Σ (xᵢ − x̄)² ]

Example (continuing from above):

σ = √4 = 2

s = √4.57 = 2.14

What does it mean?

If mean = 70 and std = 10:

Most data falls between 60 and 80 (mean ± 1 std)

Almost all data falls between 50 and 90 (mean ± 2 std)


The Empirical Rule (68-95-99.7 rule) for normal distribution:

mean ± 1σ  →  68% of data falls here

mean ± 2σ  →  95% of data falls here

mean ± 3σ  →  99.7% of data falls here

## Uses of Statistics in Data Processing



Statistics is used at every step of the data pipeline:
1. Handling missing values:

Fill missing values with mean (for symmetric data) or median (for skewed data)

2. Detecting outliers:

A value beyond mean ± 3σ is usually an outlier
outlier if |x − μ| > 3σ

3. Feature scaling / normalization:
Min-Max Scaling:

x_scaled = (x − min) / (max − min)
Scales to range [0, 1]

Z-score Standardization:

z = (x − μ) / σ
Centers at 0 with std = 1. Uses mean and std directly.

4. Understanding data distribution:

Check if data is symmetric or skewed using mean, median, mode comparison.

5. Correlation:

r = Σ(xᵢ−x̄)(yᵢ−ȳ) / √[Σ(xᵢ−x̄)² × Σ(yᵢ−ȳ)²]
Measures linear relationship between two variables. Range: −1 to +1.

## Probability

BASIC PROBABILITY

  P(E) = n(E) / n(S)

  0 ≤ P(E) ≤ 1

COMPLEMENT

  P(E') = 1 − P(E)

SUM RULE

  Mutually exclusive:      P(A∪B) = P(A) + P(B)

  Not mutually exclusive:  P(A∪B) = P(A) + P(B) − P(A∩B)

PROBABILITY TREE

  Along path  →  multiply

  Between paths → add

CONDITIONAL PROBABILITY

  P(A|B) = P(A∩B) / P(B)

MULTIPLICATION RULE

  P(A∩B) = P(A) × P(B|A)

  Independent: P(A∩B) = P(A) × P(B)

BAYES THEOREM

  P(A|B) = P(B|A) × P(A) / P(B)
