Machine Learning is a branch of Artificial Intelligence where we teach computers to learn from data instead of explicitly programming rules. The computer finds patterns on its own and improves with experience.

Traditional Programming: Data + Rules → Output

Machine Learning: Data + Output → Rules (learned automatically)

Types of Machine Learning
There are 3 main types. The key difference is how the model learns — with labels, without labels, or through rewards.

Type 1 — Supervised Learning

The model is trained on labeled data — every input has a known correct answer. 
It learns the mapping input → output.
Two sub-types: Classification (predict a category) and Regression (predict a number).

Type 2 — Unsupervised Learning

No labels — the model finds hidden structure in raw data on its own. Two main tasks: Clustering (group similar things) and Dimensionality Reduction (simplify data).



Type 3 — Reinforcement Learning

An agent takes actions in an environment, receives rewards or penalties, and learns to maximize its total reward over time. No dataset — it learns by doing.

Exploratory Data Analysis (EDA)

EDA is the process of examining and understanding a dataset before building any model. You explore the data visually and statistically to find patterns, spot anomalies, check assumptions, and understand relationships between variables.

Think of it as a doctor examining a patient before prescribing medicine — you need to understand the data fully before doing anything with it.

After training an ML model, we need to measure how good or bad its predictions are. Different metrics tell us different things — no single metric tells the full story.

First, understand the foundation — the Confusion Matrix.

Classification Metrics
These metrics all come directly from the confusion matrix values: TP, TN, FP, FN.

Precision vs Recall Trade-off

These two always pull in opposite directions — improving one usually hurts the other.

Regression Metrics

When your output is a number (not a category), you use these to measure how far off your predictions are.
