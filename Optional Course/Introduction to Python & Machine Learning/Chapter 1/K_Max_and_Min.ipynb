```python
num1, num2, num3 = map(int,input().split())

if num1 >= num2 and num1 >= num3:
  maxNum = num1
elif num2 >= num1 and num2 >= num3:
  maxNum = num2
else:
  maxNum = num3
  
if num1 <= num2 and num1 <= num3:
  minNum = num1
elif num2 <= num1 and num2 <= num3:
  minNum = num2
else:
  minNum = num3
  
print(minNum, maxNum)
```
