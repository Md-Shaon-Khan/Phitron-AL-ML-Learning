Operators in Python

Arithmatic: + - * / % 

```python
x = 10
y = 3
```

```python
sum = x + y
sum
```

```python
sub = x - y
sub
```

```python
mult = x * y
mult
```

```python
div = x / y
div
```

```python
rem = x % y
rem
```

```python
print(type(sum), type(sub), type(mult), type(div), type(rem))
```

```python
type(sum), type(sub), type(mult), type(div), type(rem)
```

Comparison

> >= < <= != ==

```python
x = 10
y = 3
```

```python
greater_than = x > y

greater_than
```

```python
greater_than_or_equal = 10 >= 10

greater_than_or_equal, type(greater_than_or_equal)
```

```python
equal_or_not = 10 == 10

equal_or_not, type(equal_or_not)
```

```python
euqal_or_not = 10 != 10

euqal_or_not, type(euqal_or_not)
```

```python
less_than = x < y
less_than_equal = x <= 10
```

```python
less_than, less_than_equal
```

```python
equal = 10 != None
equal
```
