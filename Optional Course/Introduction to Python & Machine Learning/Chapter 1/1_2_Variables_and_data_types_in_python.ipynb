```python
#Rule of variable in python
#1. should be only english letter, number and underscore
#2. should not start with number
#3. should not be a keyword
#4. should not contain space
#5. should be meaningful
#Example of variable
```

```python
phone_number = 1633040670 #python don't allow leading zero in number

print(phone_number)
print(type(phone_number))
print(f"{phone_number:011d}")
print(type(f"{phone_number:011d}"))
```
