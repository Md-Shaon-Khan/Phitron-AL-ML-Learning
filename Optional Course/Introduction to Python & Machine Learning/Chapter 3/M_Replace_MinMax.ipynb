```python
n = int(input())

list1 = list(map(int,input().split()))

minimum = min(list1)
maximum = max(list1)
newList = []
#if it is a string then
# newList = list1.replace(minimum,maximum)
# newList = list1.replace(maximum,minimum)

for i in range(n):
    if list1[i] == minimum:
        newList.append(maximum)
    elif list1[i] == maximum:
        newList.append(minimum)
    else:
        newList.append(list1[i])


print(*newList)
```
