```python
n = int(input())

arr = list(map(int,input().split()))

even = 0
odd = 0
pos = 0
neg = 0

for a in range(0,len(arr)):
  if arr[a] % 2 == 0:
    even += 1
  else:
    odd += 1
  if arr[a] > 0:
    pos += 1
  elif arr[a] < 0:
    neg += 1

print(f"Even: {even}")
print(f"Odd: {odd}")
print(f"Positive: {pos}")
print(f"Negative: {neg}")
```
