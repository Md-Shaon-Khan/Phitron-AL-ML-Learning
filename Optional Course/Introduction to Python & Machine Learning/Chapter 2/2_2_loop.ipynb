Loop in Python

```python
for i in range(10):
  print(i,"Hello World!")
```

```python
list(range(10))
```

```python
for i in range(1,10):
  print(i,"Hello World!")
```

```python
for i in range(1,10,3):
  print(i,"Hello World!")
```

```python
for i in range(1,5):
  for j in range(6,8):
    print(i,j)
```
