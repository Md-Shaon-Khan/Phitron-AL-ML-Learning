```python
sum = 0

for i in range(1,11):
  if i % 2 == 0:
    sum += i
    print(i, "is added to the sum.")

print(sum)
```

```python
sum1 = 0

for i in range(1,11):
  print(i," is processing...")
  if i % 2 == 0:
    continue
  
  sum1 += i
  print(i," is added.")

print(sum1)
```

```python
accuracy = 90

for i in range(100):
  accuracy += 1
  print(i, accuracy)
  
  if accuracy == 100:
    break
```

```python
acc = 95

while True:
  if acc == 100:
    break
  acc += 1
  print(acc)

print(acc)
```
