```python
import numpy as np
```

```python
array_1 = np.array([1, 2, 3, 4, 5])

array_2 = np.array([[1, 2, 3],
                    [7, 5, 1]])

array_3 = np.array([
  [[1, 2, 3],[4, 5, 6]],
  [[1, 6, 2],[3, 5, 7]],
  [[9, 8, 1],[4, 6, 1]]
])
```

```python
print(array_1.ndim)
print(array_2.ndim)
print(array_3.ndim)
```

```python
print(array_1.shape)
print(array_2.shape)
print(array_3.shape)
```

```python
print(array_1.dtype)
print(array_3.dtype)
```
