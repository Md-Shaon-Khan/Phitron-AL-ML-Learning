```python
import numpy as np
```

```python
arr = np.array([1, 2, 3])
print(arr.dtype)
```

```python
arr = np.array([1, 2, 3.4])
print(arr.dtype)
```

```python
arr = np.array([1, 2, 3.3, "Hello"])
print(arr.dtype)
```

```python
arr = np.array([1, 2, 3],dtype=np.int16)

print(arr.dtype)
print(arr)
```

```python
arr = np.array([1, 2, 3, 4, 5.5])
print(arr.dtype)

arr2 = arr.astype(np.int32)
print(arr2.dtype)
print(arr.dtype)
print(arr2)
```

```python
arr = np.array([1, 2, 3.2, "Hello"],dtype=np.float64)
arr
```
