```python
import numpy as np

lst = [10, 20, 30, 40, 50]

array = np.array(lst,dtype=np.int32)
array, array.dtype
```

```python
matrix = [[1, 2, 3],
          [4, 5, 6]]

arr = np.array(matrix)

arr, arr.dtype, arr.ndim, arr.shape
```

```python
tpl = (10, 20, 30)

arr = np.array(tpl, dtype=np.int32)

arr
```

```python
st = {1, 2, 3, 3}

arr = np.array(st, dtype = np.int32)

arr
```

```python
st = {1, 2, 3, 3}

arr = np.array(list(st), dtype = np.int32)

arr
```

```python
dc = {
  'a': 10,
  'b': 20,
  'c': 30
}

arr = np.array(list(dc.keys()))
arr
```

```python
arr = np.array(list(dc.values()))
```

```python
arr = arr.astype(np.int16)
arr, arr.ndim
```

```python
arr = np.array(list(dc.items()))
arr, arr.dtype
```
