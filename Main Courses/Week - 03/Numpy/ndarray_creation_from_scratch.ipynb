```python
import numpy as np
```

```python
arr = np.zeros((2,3), dtype = np.int8)

print(arr)
```

```python
arr2 = np.zeros_like(arr)
print(arr2)
```

```python
arr = np.ones((3, 4, 3), dtype = np.int8)
print(arr)
```

```python
arr = np.empty((4,3), dtype=np.int8)

print(arr)
```

```python
arr2 = np.empty_like(arr)

print(arr2)
```

```python
arr = np.full((3,2), np.inf)

print(arr, arr.dtype)
```

```python
arr2 = np.full_like(arr, np.inf, dtype = np.float64)

print(arr2, arr2.dtype)
```

```python
arr = np.array([[1,2,3],
                [4,5,6]])

print(arr)
```

```python
arr2 = np.copy(arr)

print(arr2)
```
