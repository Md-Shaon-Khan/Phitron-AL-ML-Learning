```python
import numpy as np
```

```python
arr1 = np.array([1,2,3])
arr2 = 5

print(arr1 + arr2)
```

```python
A = np.array([[1,2,3],[4,5,6]])
B = np.array([10,20,30])

print(A+B)
```

```python
A = np.array([
  [[1],[2],[3]]
])
B = np.array([[10, 20, 30]])

C = A + B

print(A.shape)
print(B.shape)
print(C.shape)

print(C)
```

```python
arr = np.array([1,2,3])

print(arr.shape)

arr_col = arr[:, np.newaxis]
print(arr_col.shape)
```

```python
arr = np.array([1,2,3])

new_arr = np.expand_dims(arr, axis = 0)
print(new_arr.shape)

new_arr = np.expand_dims(arr, axis = 1)
print(new_arr.shape)
```

```python
arr = np.array([1, 2, 3])

result = np.broadcast_to(arr, (3,3))
print(result)
```

```python
titled = np.tile(arr, (2,3))
print(titled)
```

```python
X = np.array([[10,20,30],[40,50,60]])

mean = np.mean(X, axis=0)
print(mean)
normalized = X - mean
print(normalized)
```

```python
X = np.array([[1,2,3],[4,5,6]])

scales = np.array([1,10,100])
scaled = X * scales
print(scaled)
```
