```python
import numpy as np
```

```python
arr = np.array([10,20,30,40])

print(arr > 25)
print(arr[arr > 25])
```

```python
result = np.logical_and(arr > 10, arr < 35)
print(result)
print(arr[result])
```

```python
result = np.logical_or(arr < 15, arr > 35)
print(result)
```

```python
condition = arr > 15
print(condition)

print(np.logical_not(condition))
```

```python
x = np.array([True, False, True, False])
y = np.array([True, True, False, False])
print(np.logical_xor(x, y))
```

```python
arr = np.array([0, 0, 1, 0])
print(np.any(arr))

arr = np.array([0, 0, 0, 0])
print(np.any(arr))

arr = np.array([0, 0, -1, 0])
print(np.any(arr))
```

```python
arr = np.array([0, 0, 1, 0])
print(np.all(arr))

arr = np.array([2, 22, 1, -9])
print(np.any(arr))
```

```python
arr = np.array([5, 10, 15, 20])
result = (arr > 5) & (arr < 18)
print(result)
print(arr[result])
```

```python
import numpy as np

data = np.array([45, 50, np.nan, 60, 30])
valid = np.logical_and(~np.isnan(data), data >= 50)
filtered_data = data[valid]
print(filtered_data)
```

```python
import numpy as np

data = np.array([45, 50, np.nan, 60, 30])
valid = np.logical_and(np.logical_not(np.isnan(data)), data >= 50)
filtered_data = data[valid]
print(filtered_data)
```
