```python
arr = np.array([
  [12,8,10],
  [3,1,8]
])

print(arr)

np.sort(arr, axis = 1)
```

```python
arr = np.array([10,2,7,1])

desc_sorted = np.sort(arr)[::-1]

desc_sorted
```

```python
arr = np.array([50, 10, 30, 20])

indices = np.argsort(arr)
print(indices)
print(arr[indices])
```

```python
data = np.array([("Alice", 25), ("Bob", 30), ("Charlie", 20)], dtype=[('name', 'U10'), ('age', 'i4')])
print(np.sort(data, order='age'))
```

```python
arr = np.array([9, 4, 7, 1, 3, 8])
part = np.partition(arr, 4)
print(part)
```

```python
scores = np.array([80, 95, 70, 60, 85])
rank = np.argsort(scores)[::-1]  # descending order
print("Rank Order:", rank)
print("Top 3 Scores:", scores[rank][:3])
```
