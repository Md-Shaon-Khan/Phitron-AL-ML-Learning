```python
import numpy as np
```

```python
arr = np.array([10, 20, 30, 40, 50])
indices = np.where(arr > 25)
print(indices)
print(arr[indices])
```

```python
new_arr = np.where(arr > 30, 1984, 0)
print(new_arr)
```

```python
arr = np.array([3, 7, 3, 2, 9, 3])
indices = np.where(arr == 3)
print(indices)
```

```python
arr = np.array([0, 2, 0, 5, 7, 0])
print(np.nonzero(arr))
```

```python
arr = np.array([12, 45, 7, 22, 50])
print(np.argmax(arr)) # index of max value
print(np.argmin(arr))# index of min value
```

```python
arr = np.array([10, 20, 30, 40, 50])

pos = np.searchsorted(arr, 35)
print(pos)
```

```python
arr = np.array([10, 20, 30, 30, 40])
print(np.searchsorted(arr, 30, side='left'))
print(np.searchsorted(arr, 30, side='right'))
```

```python
arr = np.arange(10)

condition = arr % 2 == 0

result = np.extract(condition, arr)
print(result)
```

```python
scores = np.array([55, 72, 48, 90, 67])
passed_idx = np.where(scores >= 60)
print("Indices:", passed_idx)
print("Passed Scores:", scores[passed_idx])
```
