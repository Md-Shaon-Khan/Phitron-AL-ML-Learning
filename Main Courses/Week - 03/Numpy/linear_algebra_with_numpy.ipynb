```python
import numpy as np
```

```python
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
print(A)
print(B)
```

```python
A = np.array([[1, 2], [3, 4]])
B = np.array([[2, 0], [1, 2]])
C = np.dot(A, B)
print(C)
```

```python
C = A @ B
print(C)
```

```python
A = np.array([[1, 2, 3], [4, 5, 6]])
print(A.T)
```

```python
A = np.array([[1, 2], [3, 4]])
print(np.linalg.det(A))
```

```python
A = np.array([[1, 2], [3, 4]])
invA = np.linalg.inv(A)
print(invA)
```

```python
I = np.eye(3)
print(I)
```

```python
A = np.array([[2, 1], [1, -1]])
B = np.array([8, 2])
X = np.linalg.solve(A, B)
print(X)
```

```python
A = np.array([[4, -2], [1, 1]])
values, vectors = np.linalg.eig(A)
print("Eigenvalues:", values)
print("Eigenvectors:\\n", vectors)
```

```python
A = np.array([[3, 4]])
print(np.linalg.norm(A))
```

```python
A = np.array([[1, 2], [2, 4]])
print(np.linalg.matrix_rank(A))
```

```python
A = np.array([[2, 1], [1, 3]])
B = np.array([8, 18])
X = np.linalg.solve(A, B)
print(X)
```
