```python
import numpy as np
```

```python
A = np.array([1,2,3])

print(A.ndim, A.shape)

print('_' * 50)

A = np.array([[1,2,3],
              [4,5,6]])

print(A.ndim, A.shape)

print('_' * 50)

A = np.array([[[1,2,3],
              [4,5,6]],
              [[1,2,3],
              [4,5,6]],
              [[1,2,3],
              [4,5,6]]
              ])

print(A.ndim, A.shape)
```

```python
A = np.zeros((3,3))

print(A)

print('_' * 50)

A = np.ones((2,5))

print(A)

print('_' * 50)

A = np.ones((2,5))

print(A)

print('_' * 50)

A = np.full((3,3), 7)

print(A)
```

```python
A = np.arange(10, 50)

print(A[A % 2 == 0])
```

```python
A = np.arange(0,16).reshape((4,4))

print(A)

print('_' * 50)

print(A[::, :1])

print('_' * 50)

print(A[::, 3:])

print('_' * 50)

print(A[1:3, 1:3])
```
