```python
import numpy as np


arr1 = np.array([10, 20, 30, 40])
arr2 = np.array([1, 2, 3, 4])

print(arr1 + arr2)
print(arr1 - arr2)
print(arr1 * arr2)
print(arr1 / arr2)
```

```python
arr = np.array([5, 10, 15])

print(arr + 5)
print(arr * 2)
```

```python
arr = np.array([1, 2, 3, 4])

print(arr > 2)
print(arr == 3)
```

```python
arr = np.array([1,2,3,5])

arr += 4

print(arr)
```

```python
arr = np.array([1,2,3,4,5,9])

print(np.sum(arr))
print(np.mean(arr))
print(np.std(arr))
print(np.min(arr))
print(np.max(arr))
```

```python
angles = np.array([0, np.pi/2, np.pi])

print(np.sin(angles))
print(np.cos(angles))
print(np.tan(angles))
```

```python
arr = np.array([1, 2, 3])

print(np.exp(arr))
```

```python
print(np.log(arr))
```

```python
print(np.log10(arr))
```

```python
arr = np.array([1.2, 2.6, 3.9])

print(np.round(arr))
print(np.floor(arr))
print(np.ceil(arr))

powers = np.array([2, 3, 4])
print(np.power(powers,2))
```
