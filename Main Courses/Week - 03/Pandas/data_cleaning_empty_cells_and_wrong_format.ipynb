```python
df2 = pd.read_csv("student_performance_prediction.csv")  
mean_attendance = df2["Attendance Rate"].mean()
print("\nAttendance Rate Mean:", round(mean_attendance, 2))

df2["Attendance Rate"] = df2["Attendance Rate"].fillna(
    mean_attendance
)

print(df2["Attendance Rate"].isnull().sum())

df3 = pd.read_csv("student_performance_prediction.csv")
median_attendance = df3["Attendance Rate"].median()
df3["Attendance Rate"] = df3["Attendance Rate"].fillna(
    median_attendance
)

df3["Attendance Rate"].isnull().sum()
```
