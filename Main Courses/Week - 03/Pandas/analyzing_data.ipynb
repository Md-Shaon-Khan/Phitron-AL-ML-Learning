```python
result = df.groupby("Parent Education Level")["Previous Grades"].mean().round(2)
print(result)
```

```python
result2 = df.groupby(
    "Participation in Extracurricular Activities"
)["Study Hours per Week"].mean().round(2)
print(result2)
```

```python
result3 = df.groupby("Passed")["Study Hours per Week"].mean().round(2)
print(result3)
```

```python
result4 = df.groupby("Passed")["Previous Grades"].mean().round(2)
print(result4)
```

```python
result = df.groupby("Passed")[
    ["Study Hours per Week", "Previous Grades", "Attendance Rate"]
].agg(["mean", "max", "min"]).round(2)

print(result)
```
