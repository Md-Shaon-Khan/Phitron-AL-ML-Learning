```python
def simple_generator():
  yield 1
  yield 2
  yield 3
  
gen = simple_generator()
print(next(gen))
print(next(gen))
print(next(gen))
print(next(gen))
```

```python
lst = [x for x in range(1,11)]
print(lst)

def data_loader(lst,chunk):
  for i in range(0,len(lst),chunk):
    yield lst[i:i+chunk]

x = data_loader(lst,3)
print(next(x))
print(next(x))
print(next(x))
print(next(x))
```

| বিষয়                 | map()                                   | filter()                         | reduce()                                   |
| -------------------- | --------------------------------------- | -------------------------------- | ------------------------------------------ |
| কাজ                  | প্রতিটি উপাদানের উপর function apply করে | শর্ত অনুযায়ী উপাদান বাছাই করে    | পুরো list কে একটি single value-এ কমিয়ে আনে |
| Input                | function + iterable                     | function (True/False) + iterable | function + iterable                        |
| Output               | নতুন iterable (modified values)         | filtered iterable                | single value                               |
| Function return type | যেকোনো value                            | অবশ্যই True/False                | accumulated value                          |
| Use case             | transform করা                           | condition দিয়ে filter করা        | sum, product ইত্যাদি                       |
| Module               | built-in                                | built-in                         | `functools` থেকে import করতে হয়            |


```python
z = [1,2,3,4]

result = list(map(str,z))

print(result,type(result[0]))
```

```python
z = [1,2,3,4]

result = list(map(lambda x: x%2==0,z))

print(result,type(result[0]))
```

```python
z = [1,2,3,4]

result = list(filter(lambda x:x%2==0,z))
print(result)
```

```python
from functools import reduce

z = [1,2,3,4]
result = reduce(lambda x,y:x+y,z,10)

print(result)
```

map → Modify (পরিবর্তন করে)
filter → Select (বেছে নেয়)
reduce → Combine (একটায় নামায়)
