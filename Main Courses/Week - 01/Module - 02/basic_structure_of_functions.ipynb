```python
def sum_number(*args):
  print(args)
  print(sum(args))
  return args
  
  
sum_number(10,20,30,40) , type(sum_number(10,20,30,40))
```

```python
def add_args_number(*args):
  countSum = 0
  
  for lst in args:
    for i in lst:
      countSum += i
  
  print(countSum)
 
add_args_number([1,2,3,4,5],[10,20,30],[1],[1,2,5,6,8,1,1]) 
  
```

```python
lst = [x**2 for x in range(500) ]

print(lst)
```

```python
def data_loader(chunkk_sizee,lst):
  for i in range(len(lst)):
    yield lst[i:i+chunkk_sizee]
    
    
x = data_loader(5,lst)
print(type(x))
print(next(x))
print(next(x))
```

```python
def data_loader(chunkk_sizee,lst):
  for i in range(0,len(lst),chunkk_sizee):
    yield lst[i:i+chunkk_sizee]
    
    
x = data_loader(5,lst)
print(type(x))
print(next(x))
print(next(x))
```

```python
import pandas as pd

df = pd.DataFrame({'Temperature (F)':[32,212,98.6,68]})

df['Temperature (C)'] = df['Temperature (F)'].apply(lambda x: (x - 32) * (5/9))

df
```

```python
df = pd.DataFrame({'A':[1,2,3,4,5,9],
                   'B': [5,6,7,8,9,99]})

df['A'] = df.A.map(lambda x: x ** 0.5)
df 
```
