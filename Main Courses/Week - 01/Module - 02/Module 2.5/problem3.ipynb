```python
lst = [1, 2, 3, 4]

lstSquare = lambda x: x ** 2
 
for x in lst:
    print(lstSquare(x),end=" ")
```

```python
def power(base, exponent = 2):
  return base ** exponent

full = input()
base = full.split()
#print(base)
if len(base) > 1:
  
  print(power(int(base[0]),int(base[1])))
else:
  print(power(int(base[0])))
```

```python
def power2(base,exponent=2):
  return base ** exponent

parts = list(map(int,input().split()))
print(power2(*parts))
```
