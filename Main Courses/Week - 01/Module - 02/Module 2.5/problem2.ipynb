```python
lst = list(map(int,input('data = ').split(', ')))
batch = int(input('batch_size = '))


def batch_generator(lst,batch):
  for i in range(0, len(lst), batch):
    yield lst[i:i + batch]
    

x = batch_generator(lst,batch)

count = len(lst)

while count >= 0:
  print(next(x))
  count -= batch
```
