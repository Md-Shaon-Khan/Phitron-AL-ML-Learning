```python
with open('emails.txt','r') as file:
  #print(file.read())
  emails = file.readlines()
  
cleaned = sorted(set(email.strip().lower() for email in emails))
  
with open('unique_emails.txt','w') as f:
  for email in cleaned:
    f.write(email+'\n')
  
```

```python
with open('names.txt','r') as file:
  names = file.readlines()
  print(names)

cleaned = sorted(set(name.capitalize().strip() for name in names))

with open('clean_names.txt','w') as filee:
  for namee in cleaned:
    filee.write(namee+'\n')
```

```python
numbers = [10,20,30,40,50]

print(numbers)

try:
  index = int(input("Enter an index number: "))
  print(f"Value at index {index}, is {numbers[index]}")
except IndexError:
  print("Invalid Index")
except ValueError:
  print("Enter a valid Integer!")
except Exception as e:
  print(e)
```
