```python
from functools import reduce

with open('./File/sample.txt','r') as fil:
  rl = fil.readlines()
  print(len(rl))
  
  number_of_words = list(map(lambda x: len(x.split()),rl))
  tw = reduce(lambda x,y:x+y, number_of_words)
  print(tw)
  
  sl = list(map(str.strip,rl))
  sl = list(map(lambda x:x.replace(" ",""),sl))
  print(sl)
  noc = list(map(lambda x:len(x),sl))
  tnoc = reduce(lambda x,y:x+y,noc)
  print(tnoc)
  
```
