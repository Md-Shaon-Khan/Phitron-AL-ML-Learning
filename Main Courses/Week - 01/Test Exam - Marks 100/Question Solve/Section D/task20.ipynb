```python
sentence = input()

words = sentence.split()
word_length = {}

for word in words:
    word_length[word] = len(word)

print(word_length)


sorted_dict = dict(sorted(word_length.items(), key=lambda x: x[1], reverse=True))


longest_word = next(iter(sorted_dict))

print("Longest word:", longest_word)
```
