```python
number = int(input("Enter a number: ").strip())

if (number % 2 == 0) and (number % 3 == 0 and number % 5 == 0):
  print("Number is even and also divisible by both 3 and 5.")
elif (number % 2 == 0) and (number % 3 != 0 or number % 5 != 0):
  print("Number is only even.")
elif (number % 2 == 1) and (number % 3 == 0 and number % 5 == 0):
  print("Number is odd and also divisible by both 3 and 5.")
else:
  print("Number is only odd.")
```
