```python
subject1 = int(input("Enter first subject marks: ").strip())
subject2 = int(input("Enter second subject marks: ").strip())
subject3 = int(input("Enter third subject marks: ").strip())

average_marks = (subject1 + subject2 + subject3) / 3

if average_marks >= 80:
  print("A+")
elif average_marks >= 70:
  print("A")
elif average_marks >= 60:
  print("B")
elif average_marks >= 50:
  print("A")
elif average_marks < 50:
  print("Fail")
```
