```python
def create_file(filename,content):
  with open(filename,'w') as file:
    file.write(content)
  with open(filename,'r') as filee:
    print(filee.read())

filename = input()
content = input()

create_file(filename,content)
```

```python
def create_file(filename, content):
    with open(filename, 'w') as file:
        file.write(content)

filename, content = input().split(",", 1)

create_file(filename.strip().strip('"'), content.strip().strip('"'))
```

```python
def read_lines(filename):
  with open(filename,'r') as file:
    lines = file.readlines()
    i = 1
    for line in lines:
      print(f"{i}: {line}",end='')
      i += 1

read_lines('data.txt')
```

```python
def add_log(message):
  with open('log.txt','a') as file:
    file.write(f"{message}\n")

message = input()
add_log(message)
```

```python
def read_from_position(filename,position):
  try:
    with open(filename,'r') as file:
      file.seek(position)
      print(file.read())
  except FileNotFoundError:
    print("No file founded in this name.")
    
filename = input("filename = ")
position = int(input())

read_from_position(filename,position)
```

```python
def safe_read(filename):
  try:
    with open(filename,'r') as file:
      note = file.read()
  except FileNotFoundError:
    print("Error: File not found.")
  except Exception as e:
    print(e)
  else:
    print(note.strip())
  finally:
    print("Operation complete")

safe_read("logg.txt")
```
