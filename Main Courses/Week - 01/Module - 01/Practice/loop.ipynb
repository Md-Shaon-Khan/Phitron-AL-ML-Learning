```python
def digit_sum(n):
  
  sumTotal = 0
  while n > 0:
    ext = n % 10
    sumTotal += ext
    n = n // 10
  
  return sumTotal

inpNum = int(input())
sumT = digit_sum(inpNum)
print(sumT)
```

```python
def print_table(n):
  for i in range(1,11):
    print(n,' x ', i,' = ',n*i)

n = int(input())
print_table(n)
```

```python
def FizzBuzz(n):
  for i in range(1,n+1):
    if (i % 5 == 0) and (i % 3 == 0):
      print('FizzBuzz')
    elif (i % 3 == 0):
      print('Fizz')
    elif (i % 5 == 0):
      print('Buzz')
    else:
      print(i)

n = int(input())
FizzBuzz(n)  
```
