```python
def remove_duplicates(lst):
  n = len(lst)
  i = 0
  while i < len(lst):
    j = i + 1
    while j < len(lst):
      if lst[i] == lst[j]:
        lst.pop(j)
      else:
        j += 1
    i += 1
  return lst

lst = list(map(int,input().split(',')))
lst1 = remove_duplicates(lst)
print(lst1)
```
