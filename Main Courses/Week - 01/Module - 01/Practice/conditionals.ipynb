```python
def is_even(n):
  if n % 2 == 0:
    return 'Even'
  else:
    return 'Odd'
  
n = int(input())
string = is_even(n)
print(string)
```

```python
def grade(score):
  if score >= 90 and score <= 100:
    return 'A'
  elif score >= 80 and score <= 89:
    return 'B'
  elif score >= 70 and score <= 79:
    return 'C'
  elif score >= 60 and score <= 60:
    return 'D'
  elif score < 60:
    return 'F'
  
score = int(input())
sco = grade(score)
print(sco)
```
