```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
student = pd.read_csv('student_data.csv')
student.head()
```

```python
sns.scatterplot(
    data = student,
    x = 'study_hours',
    y='test_score',
    hue='gender'
)

plt.show()
```

```python
sns.relplot(
    kind = 'scatter',
    data = student,
    x = 'study_hours',
    y = 'test_score',
    hue = 'gender'
    
)
```

```python
sns.relplot(
    kind = 'scatter',
    data = student,
    x = 'study_hours',
    y = 'test_score',
    col = 'gender'
)
```

```python
sns.relplot(
    kind = 'scatter',
    data = student,
    x = 'study_hours',
    y = 'test_score',
    col = 'gender',
    row = 'hostel'
)
```

```python
sns.relplot(
    kind = 'scatter',
    data = student,
    x = 'study_hours',
    y = 'test_score',
    col = 'hostel',
    row = 'gender'
)
```

```python
sns.relplot(
    kind = 'scatter',
    data = student,
    x = 'study_hours',
    y = 'test_score',
    col = 'week',
    col_wrap = 2
)
```
