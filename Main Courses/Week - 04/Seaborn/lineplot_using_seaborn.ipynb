```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
student = pd.read_csv('student_data.csv')
student.head()
```

```python
sns.lineplot(data=student, x='week', y='attendance_rate',errorbar=None)
plt.show()
```

```python
sns.lineplot(data=student, x='week', y='attendance_rate')
plt.show()
```

```python
sns.lineplot(data = student, x = 'week', y = 'attendance_rate', hue = 'gender', errorbar = None)
plt.show() 
```

```python
sns.lineplot(data = student, x = 'week', y = 'attendance_rate', hue = 'class_level', errorbar = None)
plt.show() 
```

```python
sns.relplot(kind='line', data=student, x='week',y='attendance_rate',hue='gender',errorbar=None)
```

```python
sns.relplot(
    kind='line',
    data=student,
    x='week',
    y='attendance_rate',
    hue='gender',
    col='gender',  
    errorbar=None
)
```

```python
sns.relplot(
    kind='line',
    data=student,
    x='week',
    y='attendance_rate',
    hue='gender',
    col='class_level',
    errorbar=None
)
```

```python
sns.relplot(
    kind = 'line',
    data = student,
    x = 'week',
    y = 'test_score',
    hue = 'class_level',
    col = 'gender',
    errorbar = None
)
```
