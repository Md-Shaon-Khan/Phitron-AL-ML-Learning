```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
student = pd.read_csv('student_data.csv')
student
```

```python
sns.regplot(
    data = student,
    x = 'study_hours',
    y = 'test_score'
)

plt.show()
```

```python
sns.lmplot(
    data = student,
    x = 'study_hours',
    y = 'test_score',
    hue = 'gender'
)
```
