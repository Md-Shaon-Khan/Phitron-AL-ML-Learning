```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
student = pd.read_csv('student_data.csv')
student
```

```python
sns.countplot(
    data = student,
    x = 'gender'
)
```

```python
sns.countplot(
    data = student,
    x = 'subject',
    hue = 'gender'
)
```

```python
sns.barplot(
    data=student,
    x='gender',
    y='test_score',
    errorbar=None
)
```

```python
sns.barplot(
    data = student,
    x = 'gender',
    y = 'test_score',
    errorbar = None,
    estimator = np.median
)
```

```python
sns.barplot(
    data = student,
    x = 'gender',
    y = 'test_score',
    errorbar = None,
    estimator = np.max
)
```

```python
sns.barplot(
    data = student,
    x = 'gender',
    y = 'test_score',
    errorbar = None,
    estimator = np.min
)
```
