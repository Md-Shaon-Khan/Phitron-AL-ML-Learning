```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
student = pd.read_csv('student_data.csv')
student
```

```python
sns.scatterplot(
    data = student,
    x = 'study_hours',
    y = 'test_score'
)
plt.show
```

```python
sns.scatterplot(
    data = student,
    x = 'study_hours',
    y = 'test_score',
    hue = 'gender',
    style = 'subject',
    size = 'Tshirt_size'
)

plt.show()
```

```python
sns.relplot(
    kind = 'scatter',
    data = student,
    x = 'study_hours',
    y = "test_score",
    hue = 'gender',
    style = 'subject',
    size = 'Tshirt_size'
)
```
