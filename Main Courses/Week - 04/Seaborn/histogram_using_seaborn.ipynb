```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
student = pd.read_csv('student_data.csv')
student
```

```python
sns.histplot(
    data = student,
    x = 'attendance_rate',
    hue = 'gender',
    bins = 10,
    element = 'step'   
)

plt.show()
```

```python
sns.displot(
    kind = 'hist',
    data = student,
    x = 'attendance_rate',
    bins = 6
)
```

```python
sns.displot(
    kind = 'hist',
    data = student,
    x = 'attendance_rate',
    col = 'gender',
    bins = 6
)
```

```python
sns.displot(
    kind = 'hist',
    data = student,
    x = 'study_hours',
    col = 'gender'
)
```
