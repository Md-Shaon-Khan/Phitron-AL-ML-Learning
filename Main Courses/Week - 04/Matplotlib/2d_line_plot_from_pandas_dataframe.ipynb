```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Example 1: Creating DataFrame manually
data = {
'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
'Sales_A': [220, 270, 300, 350, 400, 420],
'Sales_B': [180, 230, 280, 310, 360, 390]
}
df = pd.DataFrame(data)

# Example 2: Reading from CSV file
# df = pd.read_csv('sales_data.csv')
```

```python
df
```

```python
df.plot(x = 'Month', y = ['Sales_A','Sales_B'], kind = 'line', marker = 'o', linewidth = 2)
plt.grid(True)
```

```python
plt.title('Monthly Sales Comparison between Product A and B')
plt.xlabel('Month')
plt.ylabel('Sales (in Units)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(['Sales_A', 'Sales_B'], loc='upper left')
plt.tight_layout()
```

```python
df.plot(x = 'Month', y = ['Sales_A','Sales_B'], kind = 'line', marker = 'o', linewidth = 2)
plt.title('Monthly Sales Comparison between Product A and B')
plt.xlabel('Month')
plt.ylabel('Sales (in Units)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(['Sales_A', 'Sales_B'], loc='upper left')
plt.tight_layout()
```

```python
plt.style.use('seaborn-v0_8-darkgrid')

ax = df.plot(x='Month', y=['Sales_A', 'Sales_B'], marker='o', linewidth=3)
ax.set_title('Sales Performance - Styled Plot', fontsize=14, fontweight='bold')
ax.set_xlabel('Month', fontsize=12)
ax.set_ylabel('Sales (Units)', fontsize=12)
ax.legend(['Sales_A', 'Sales_B'], title='Product')
plt.show()
```
