```python
import matplotlib.pyplot as plt
```

```python
x = [10, 20, 30, 40, 50, 60]
y = [15, 25, 35, 40, 52, 60]
```

```python
plt.scatter(x, y)
plt.title('Basic Scatter Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.grid(True)
plt.show()
```

```python
plt.scatter(x, y, color = 'blue', s = 100, alpha = 0.7, edgecolors = 'black', marker='o')
plt.title('Customized Scatter plot example')
plt.xlabel("Indep. var")
plt.ylabel('Depen. Var')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
x1 = [10, 20, 30, 40, 50]
y1 = [15, 25, 32, 40, 48]

x2 = [10, 20, 30, 40, 50]
y2 = [20, 22, 28, 38, 45]

plt.scatter(x1, y1, color='red', marker='o', label='Group A')
plt.scatter(x2, y2, color='green', marker='s', label='Group B')

plt.title('Scatter Plot - Comparing Two Groups')
plt.xlabel('X Axis - Input Values')
plt.ylabel('Y Axis - Output Values')
plt.legend()
plt.grid(True)
plt.show()
```
