```python
import matplotlib.pyplot as plt
```

```python
temperature = [22, 24, 26, 28, 30, 32, 34, 36]
ice_cream_sales = [110, 130, 150, 170, 200, 230, 260, 290]


plt.scatter(temperature, ice_cream_sales, color='orange', s=80, alpha=0.8)
plt.title('Temperature vs Ice Cream Sales')
plt.xlabel('Temperature (°C)')
plt.ylabel('Sales (Units)')
plt.grid(True)
plt.show()
```

```python
plt.scatter(temperature, ice_cream_sales, color='orange', s = 100, alpha=0.99)
plt.title('Temperature vs Ice Cream Sales')
plt.xlabel('Temperature (°C)')
plt.ylabel('Sales (Units)')
plt.grid(True)
plt.show()
```
