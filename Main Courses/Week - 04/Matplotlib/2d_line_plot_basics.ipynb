```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [1, 3, 5, 7, 9]
```

```python
plt.plot(x,y1, color = 'red', linestyle = '-', marker = 'x',label = 'Line 1')
plt.plot(x,y2, color = 'green', linestyle = '--', marker = 's', label = 'Line 2')
plt.legend()
plt.show()
```

```python
plt.title('2D Line Plot Basics')
plt.xlabel('X Axis - Values')
plt.ylabel('Y Axis - Measurements')
plt.legend()
```

```python
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 19]
y2 = [1, 3, 5, 7, 15]

plt.plot(x, y1, color = 'blue', linestyle = '--', marker = 'o', label = 'Line 1')
plt.plot(x, y2, color = 'red', linestyle = '-', marker = 'x', label = 'Line 2')
plt.title('2D Line Plot')
plt.xlabel('X Axis - values')
plt.ylabel('Y Axis - Measurements')
plt.legend()
plt.grid(True)
plt.show()
```
