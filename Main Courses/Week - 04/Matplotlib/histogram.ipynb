```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
np.random.seed(10)
```

```python
data = np.random.normal(50, 10, 200)
```

```python
plt.hist(data, bins = 10, color='black', edgecolor='red')
plt.title('Histogram')
plt.xlabel('Value Range')
plt.ylabel('Frequency')
plt.grid(axis='y',alpha=0.60)
plt.show()
```

```python
plt.hist(data, bins=20, color='orange', edgecolor='black')
plt.title('Histogram with 20 Bins')
plt.xlabel('Value Range')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.7)
plt.show()
```

```python
plt.style.use('seaborn-v0_8-darkgrid')
plt.hist(data, bins=15, color='black', edgecolor='red', alpha=0.8)
plt.title('Styled Histogram Example')
plt.xlabel('Data Range')
plt.ylabel('Count')
plt.show()
```
