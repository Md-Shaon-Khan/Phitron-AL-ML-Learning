```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 11]
y2 = [1, 3, 4, 5, 9]
```

```python
plt.plot(x, y1, color='blue',label='Blue Line')
plt.plot(x, y2, color='red',label='Red Line')
```

```python
plt.plot(x, y1, color='blue',linestyle='-', label='Solid Line')
plt.plot(x, y2, color='red',linestyle='--', label='Dashed Line')
```

```python
plt.plot(x, y1, color='blue', marker='^', label='Blue Line')
plt.plot(x, y2, color='red', marker='D', label='Red Line')
```

```python
plt.plot(x, y1, color='blue', linestyle='-', marker='o', label='Line 1')
plt.plot(x, y2, color='green', linestyle='--', marker='s', label='Line 2')

plt.title('Customized Line Plot Example')
plt.xlabel('X Axis - Values')
plt.ylabel('Y Axis - Measurements')
plt.legend()
plt.grid(True)
plt.show()
```
