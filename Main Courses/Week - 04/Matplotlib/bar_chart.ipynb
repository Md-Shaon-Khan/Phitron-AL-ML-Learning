```python
import matplotlib.pyplot as plt
```

```python
categories = ['A', 'B', 'C', 'D', 'E']
values = [25, 40, 30, 20, 35]
```

```python
plt.bar(categories, values, color='skyblue', edgecolor='black')
plt.title('Basic Bar Chart')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.grid(axis='y', alpha=0.6)
plt.show()
```

```python
plt.barh(categories, values, color='orange', edgecolor='black')
plt.title('Horizontal Bar Chart')
plt.xlabel('Values')
plt.ylabel('Categories')
plt.grid(axis='x', alpha=0.6)
plt.show()
```

```python
import numpy as np

x = np.arange(len(categories))
value1 = [25, 40, 30, 20, 35]
value2 = [30, 35, 25, 28, 40]

plt.bar(x - 0.2, value1, width=0.4, color='blue', label='2024')
plt.bar(x + 0.2, value2, width=0.4, color='green', label='2025')

plt.xticks(x, categories)
plt.title('Comparison Bar Chart')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.legend()
plt.grid(axis='y', alpha=0.5)
plt.show()
```
