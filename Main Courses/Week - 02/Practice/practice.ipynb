```python
class Person:
  def __init__(self, name, age):
    self.name = name
    self.age = age
  
  def greet(self):
    print(f"Hello, my name is {self.name}")
```

```python
class Person:
  def __init__(self, name, age):
    self.name = name
    self.age = age
    
  def __str__(self):
    return f"{self.name} ({self.age})"
  
p1 = Person("Shaon Khan", 21)

print(p1)
```

```python
class Person:
  def __init__(self, fname, lname):
    self.firstname = fname
    self.lastname = lname
  
  def printname(self):
    print(self.firstname, self.lastname)
    

x = Person("John",'Doe')
x.printname()
```
