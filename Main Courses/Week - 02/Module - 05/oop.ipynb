Object-Oriented Programming (OOP)

Definition:
Object-Oriented Programming (OOP) হলো একটি programming paradigm যেখানে program design করা হয় Object এবং Class এর উপর ভিত্তি করে। এখানে data (attribute) এবং function (Method) একসাথে একটি unit হিসেবে সংগঠিত থাকে।

মৌলিক ধারণা
Class: একটি নকশা (blueprint) যা দিয়ে Object তৈরি করা হয়
Object: Class-এর একটি বাস্তব instance
Attribute: Object-এর ডাটা
Method: Class-এর ভিতরের function যা Object-এর উপর কাজ করে

```python
class MLModel:
    def __init__(self, name):
        self.name = name

    def train(self, data):
        print(self.name, "is training with data")

    def predict(self, input_data):
        print(self.name, "is predicting result")

model1 = MLModel("Heart Disease Model")
model1.train("ECG Data")
model1.predict("New ECG Signal")
```

ব্যাখ্যা

উপরের উদাহরণে:

MLModel একটি Class, যা একটি ML model-এর structure নির্ধারণ করে
model1 একটি Object, যা একটি নির্দিষ্ট ML model নির্দেশ করে
name একটি attribute, যা model-এর নাম সংরক্ষণ করে
train() এবং predict() হলো Method, যা model-এর কাজ নির্ধারণ করে
প্রয়োগ (ML ক্ষেত্রে)

Machine Learning system-এ বিভিন্ন model (যেমন ECG classifier, disease predictor) একই Class অনুসরণ করে তৈরি করা যায়। এতে code reusable হয় এবং system maintain করা সহজ হয়।


OOP একটি structured programming approach
Class থেকে Object তৈরি হয়
Object data এবং Method ধারণ করে
ML system-এ model design ও management সহজ করতে OOP ব্যবহৃত হয়

| বিষয়           | Class                                                 | Object                                                 |
| -------------- | ----------------------------------------------------- | ------------------------------------------------------ |
| সংজ্ঞা         | একটি **blueprint বা নকশা** যা দিয়ে Object তৈরি করা হয় | একটি **instance**, অর্থাৎ Class থেকে তৈরি বাস্তব জিনিস |
| প্রকৃতি        | Logical (ধারণাগত)                                     | Real (বাস্তব)                                          |
| তৈরি করার সময়  | আগে define করতে হয়                                    | Class থেকে তৈরি করা হয়                                 |
| মেমোরি ব্যবহার | নিজে আলাদা মেমোরি নেয় না (structure মাত্র)            | তৈরি হলে মেমোরি নেয়                                    |
| কাজ            | structure define করে (attribute + Method)             | actual data ধরে এবং Method execute করে                 |


```python
# ========== BASE CLASS ==========
class Student:
    institution = "Dhaka University"  # class variable

    def __init__(self, name, roll, gpa, attendance=100):
        self.name = name
        self.roll = roll
        self.gpa = gpa
        self.attendance = attendance  # percentage

    def attend_class(self, classes_missed):
        if self.attendance <= 0:
            print(f"{self.name} cannot attend — already 0%")
        else:
            self.attendance -= classes_missed
            print(f"{self.name} missed {classes_missed} classes → Attendance: {self.attendance}%")

    def study(self, hours):
        self.gpa = min(4.0, self.gpa + hours * 0.01)
        print(f"{self.name} studied {hours} hrs → GPA: {round(self.gpa, 2)}")

    def profile(self):
        print(f"\n{'='*30}")
        print(f"  Name       : {self.name}")
        print(f"  Roll       : {self.roll}")
        print(f"  GPA        : {self.gpa}")
        print(f"  Attendance : {self.attendance}%")
        print(f"  Institute  : {Student.institution}")
        print(f"{'='*30}")


# ========== CHILD CLASS (Inheritance) ==========
class ScholarStudent(Student):
    def __init__(self, name, roll, gpa, scholarship_amount):
        super().__init__(name, roll, gpa)        # parent constructor call
        self.scholarship = scholarship_amount

    def show_scholarship(self):
        print(f"{self.name} takes {self.scholarship} taka scholarship")


# ========== OBJECTS ==========
s1 = Student("Shaon", 101, 3.5)
s2 = Student("Ayan",  102, 3.2, attendance=80)
s3 = ScholarStudent("Sadia", 103, 3.8, scholarship_amount=50000)

# ========== USAGE ==========
s1.profile()
s1.attend_class(10)
s1.study(20)
s1.profile()

s3.show_scholarship()
```

```python
class Phone:
    category = "Electronics"

    def __init__(self, battery, camera, battery_percentage=100):
        self.battery = battery
        self.camera = camera
        self.battery_percentage = battery_percentage

    def charge(self, hour):
        print(f"Charge complete in {hour} hour(s)")

    def capture(self, photo):
        if self.battery_percentage == 0:
            print("No charge")
        else:
            self.battery_percentage -= photo
            print(f"Photo captured | Battery left: {self.battery_percentage}%")


class Cooling_Mechanism:
    def __init__(self, cooling_method):
        self.cooling_method = cooling_method

    def cooling_on(self):
        print(f"System cooled by: {self.cooling_method}")


class SmartPhone(Phone):
    def __init__(self, battery, camera, processor, battery_percentage=100):
        super().__init__(battery, camera, battery_percentage)
        self.processor = processor

    def charge(self, hour):
        print("Fast charging processing...")
        super().charge(hour)


class SmartPhone_CoolingMode(SmartPhone, Cooling_Mechanism):
    def __init__(self, battery, camera, processor, cooling_met, battery_percentage=100):
        SmartPhone.__init__(self, battery, camera, processor, battery_percentage)
        Cooling_Mechanism.__init__(self, cooling_met)


# ===== Object =====
pro = SmartPhone_CoolingMode("5000mAh", 50, "Snapdragon", "Liquid Cooling", 70)

print(pro.battery)            # 5000mAh
print(pro.processor)          # Snapdragon
print(pro.battery_percentage) # 70
pro.charge(1)                 # Fast charging + Charge complete in 1 hour(s)
pro.cooling_on()              # System cooled by: Liquid Cooling
pro.capture(10)               # Photo captured | Battery left: 60%
```

```python
class Camera:
  def __init__(self,camera):
    self.camera = camera
    
  def capture(self):
    print("Photo is captured.")
    

class Phone(Camera):
  def __init__(self, camera,resulation):
    super().__init__(camera)
    self.resulation = resulation
  
  def capture(self):
    print("Photo is captured by Phone.")


class DSLR(Camera):
  def __init__(self,camera,resulation):
    super().__init__(camera)
    self.resulation = resulation
  
  def capture(self):
    print("Photo is captured by DSLR.")


class Drone(Camera):
  def __init__(self,camera,resulation):
    super().__init__(camera)
    self.resulation = resulation
    
  # def capture(self):
  #   print(super().capture())
    
    
phone = Phone(15,100)
dslr = DSLR(30,200)
drone = Drone(40,400)

phone.capture()
dslr.capture()
drone.capture()
```

```python
# Parent
class Camera:
    def __init__(self, camera, resolution):
        self.camera = camera
        self.resolution = resolution

    def capture(self):
        print(f"Photo captured | Resolution: {self.resolution}MP")

    def info(self):
        print(f"Camera: {self.camera} | Resolution: {self.resolution}MP")


# Child 1
class Phone(Camera):
    def __init__(self, camera, resolution, brand):
        super().__init__(camera, resolution)
        self.brand = brand

    def capture(self):
        print(f"[{self.brand}] Phone captured | Resolution: {self.resolution}MP")


# Child 2
class DSLR(Camera):
    def __init__(self, camera, resolution, lens):
        super().__init__(camera, resolution)
        self.lens = lens

    def capture(self):
        print(f"DSLR captured with {self.lens} lens | Resolution: {self.resolution}MP")


# Child 3 — নিজের capture() নেই, parent এরটা চলবে
class Drone(Camera):
    def __init__(self, camera, resolution, altitude):
        super().__init__(camera, resolution)
        self.altitude = altitude

    def fly(self):
        print(f"Drone flying at {self.altitude}m altitude")


# ===== Objects =====
phone = Phone(15, 108, "Samsung")
dslr  = DSLR(30, 200, "50mm f/1.8")
drone = Drone(40, 400, 500)

phone.capture()   # নিজের method
dslr.capture()    # নিজের method
drone.capture()   # parent এর method (নিজের নেই)
drone.fly()       # শুধু Drone এর method

print()
phone.info()      # parent এর method — সবাই use করতে পারে
dslr.info()
drone.info()
```

## Encapsulation

```python
class Mobile:
  def __init__(self,name,model,imei):
    self.__name = name
    self.__model = model
    self.__imei = imei
    
  def charge(self):
    print("Phone is charging...")
  
  def imei_getter(self):
    return self.__imei
  
  def name_getter(self):
    return self.__name
  
  def name_setter(self,name):
    self.__name = name


phone = Mobile('Shaon','Honor 200','12345k')

phone.charge()
print(phone.imei_getter())
print(phone.name_getter())

phone.name_setter('Shaon Khan')
print(phone.name_getter())
```

```python
from abc import ABC, abstractmethod

# Abstract Class — Blueprint
class ATM(ABC):

    @abstractmethod
    def authenticate(self):
        pass

    @abstractmethod
    def withdraw(self, amount):
        pass

    @abstractmethod
    def check_balance(self):
        pass

    # Concrete method — সবার জন্য same
    def welcome(self):
        print("=" * 30)
        print("    Welcome to ATM")
        print("=" * 30)


# Child 1 — Dutch Bangla Bank
class DBBL_ATM(ATM):
    def __init__(self, account_holder, balance, pin):
        self.account_holder = account_holder
        self.balance = balance
        self.__pin = pin          # private — বাইরে থেকে দেখা যাবে না

    def authenticate(self):
        entered = int(input("Enter PIN: "))
        if entered == self.__pin:
            print(f"✅ {self.account_holder} authenticated successfully.")
            return True
        else:
            print("❌ Wrong PIN.")
            return False

    def withdraw(self, amount):
        if self.balance >= amount:
            self.balance -= amount
            print(f"✅ Withdrawn: {amount} BDT | Remaining: {self.balance} BDT")
        else:
            print("❌ Insufficient balance.")

    def check_balance(self):
        print(f"💰 Balance: {self.balance} BDT")


# Child 2 — Bkash ATM
class Bkash_ATM(ATM):
    def __init__(self, phone, balance, pin):
        self.phone = phone
        self.balance = balance
        self.__pin = pin

    def authenticate(self):
        entered = int(input("Enter bKash PIN: "))
        if entered == self.__pin:
            print(f"✅ {self.phone} verified.")
            return True
        else:
            print("❌ Wrong PIN.")
            return False

    def withdraw(self, amount):
        charge = amount * 0.018        # bKash 1.8% charge
        total = amount + charge
        if self.balance >= total:
            self.balance -= total
            print(f"✅ Withdrawn: {amount} BDT | Charge: {charge:.1f} BDT | Remaining: {self.balance:.1f} BDT")
        else:
            print("❌ Insufficient balance.")

    def check_balance(self):
        print(f"💰 bKash Balance: {self.balance} BDT")


# ===== Use =====
print("\n--- DBBL ATM ---")
dbbl = DBBL_ATM("Shaon", 10000, 1234)
dbbl.welcome()
if dbbl.authenticate():
    dbbl.check_balance()
    dbbl.withdraw(3000)
    dbbl.check_balance()

print("\n--- bKash ATM ---")
bkash = Bkash_ATM("01700000000", 5000, 5678)
bkash.welcome()
if bkash.authenticate():
    bkash.check_balance()
    bkash.withdraw(1000)
    bkash.check_balance()
```

```python
class Phone():
  def __init__(self,model,battery = 100):
    self.model = model
    self.battery = battery


class SmartPhone(Phone):
  def __init__(self, model,os, battery=100):
    super().__init__(model, battery)
    self.os = os
    
class GamingPhone(SmartPhone):
  def __init__(self, model, os,cooling_system, battery=100):
    SmartPhone.__init__(self, model, os, battery)
    self.cooling_system = cooling_system
    
  def start_game(self,name):
    self.name = name
    print(f"{self.name} on {self.model}")
    
gp = GamingPhone('HONOR 200','Android','Fan','80')
gp.start_game('Shaon')
```
