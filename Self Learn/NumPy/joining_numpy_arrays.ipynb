```python
import numpy as np

arr1 = np.array([1,2,3])
arr2 = np.array([4,5,6])

arr = np.concatenate((arr2,arr1))

arr
```

```python
arr1 = np.array([[1, 2], [3, 4]])

arr2 = np.array([[5, 6], [7, 8]])


arr = np.concatenate((arr1,arr2),axis = 1)

print(arr,arr.shape)


arrV = np.concatenate((arr2,arr1),axis=0)

arrV,arrV.shape
```
