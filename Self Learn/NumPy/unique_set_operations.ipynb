```python
import numpy as np
```

```python
arr = np.array([1,2,2,3,3,3,4])
```

```python
print(np.unique(arr))
```

```python
arr
```

```python
unique, counts = np.unique(arr, return_counts=True)
print("Unique values:", unique)
print("Counts:", counts)
print(dict(zip(unique, counts)))
```
