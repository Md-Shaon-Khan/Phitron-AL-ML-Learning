```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
visitor_data = np.array([
  [2,3],[3,3],[2,4],
  [8,7],[9,8],[8,9],
  [5,2],[6,2],[5,3]
])
```

```python
visitor_data.shape
```

```python
model = KMeans(n_clusters=3,random_state=42)

model.fit(visitor_data)
```

```python
labels = model.labels_
labels
```

```python
plt.scatter(visitor_data[:,0],visitor_data[:,1],c=labels,cmap='viridis')
plt.xlabel('Time spent')
plt.ylabel('Pages visited')
plt.title('Visitor Clustering')
plt.show()
```
