```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
gdp_per_capita = np.array([12240,27195,37675,50962,55805,]).reshape(-1,1)

life_satisfaction = np.array([4.9,5.8,6.5,7.3,7.2])
```

```python
model = LinearRegression()

model.fit(gdp_per_capita,life_satisfaction)
```

```python
new_gdp = [[22587]]

prediction = model.predict(new_gdp)
```

```python
prediction
```

```python
round(prediction[0],2)
```

```python
plt.scatter(gdp_per_capita,life_satisfaction,color='blue',label='Actual Da')
plt.plot(gdp_per_capita,model.predict(gdp_per_capita),color='red',label='linear')
plt.xlabel('GDP per capita')
plt.ylabel('Life Satisfaction')
plt.title('GDP vs Life Satisfaction')
plt.legend()
plt.show()
```
