Data preprocessing বা feature engineering-এর পরে processed data save করা দরকার। Pandas বিভিন্ন format-এ export করতে পারে। to_csv() সবচেয়ে portable, to_parquet() large dataset-এ fast ও compressed।

```python
import pandas as pd
```

```python
data = {'Name':['Alice','Bob'], 'Score':[95, 88]}

df = pd.DataFrame(data)
```

```python
df
```

```python
df.to_csv('output.csv',index=False)
```

```python
df.to_csv('output2.csv')
```

```python
df.to_csv('output1.tsv',sep=' ',index=False)
```

```python
df.to_excel('output.xlsx',sheet_name='Sheet1',index=False)
```

```python
df.to_pickle('output.pkl')
```

```python
df_back = pd.read_pickle('output.pkl')
```

```python
df_back
```

```python
df.to_json('output.json',orient='records',lines=True,indent=2)
```
