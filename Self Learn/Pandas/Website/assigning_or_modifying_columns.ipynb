```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({
  'height': [170,165,180,175],
  'weight':[70,55,90,80],
  'salary': [50000, 60000, 75000, 90000]
})
```

```python
df['bmi'] = (df['weight'] / (df['height'] / 100) ** 2).round(2)
df
```

```python
df['bmi_cat'] = pd.cut(df['bmi'], bins = [0,18.5,25,30,100], labels = ['Underweight','Normal','Overweight','Obese'])

df
```

```python
df
```

```python
df['salary_norm'] = ((df['salary'] - df.salary.min()) / (df.salary.max() - df.salary.min()))

df
```

```python
df2 = df.assign(
  height_m = lambda x: x['height'] / 100,
  is_healthy = lambda x: x['bmi'].between(18.5,24.9)
)

df2
```
