Series হলো Pandas-এর একটি one-dimensional labeled array। এটে যেকোনো ধরনের data রাখা যায় — integer, float, string, object। প্রতিটি element-এর সাথে একটি label (index) থাকে। NumPy array-এর মতো, কিন্তু custom index-এর সুবিধাসহ।

```python
import pandas as pd
import numpy as np
```

```python
s1 = pd.Series([10,20,30,40,50])

print(s1)
```

```python
s2 = pd.Series([10,20,30,40,50],index=['Math','Science','Physics','Chemistry','Biology'])

print(s2)
```

```python
s2.shape,s2.dtype,s2.index.tolist()
```

```python
data = {'age':23, 'salary':20000,'score':9.4}

s3 = pd.Series(data)
s3
```
