```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({
    'Age': [25, np.nan, 35, np.nan, 28],
    'Salary': [50000, 70000, np.nan, 60000, np.nan],
    'Score': [85, 92, 78, np.nan, 96],
    'Name': ['Alice', 'Bob', np.nan, 'David', 'Eve']
})
```

```python
df
```

```python
df
```

```python
df.isna().sum()
```

```python
(df.isna().sum() / len(df) * 100).round(1)
```

```python
df
```

```python
df['Age'] = df.Age.fillna(df.Age.mean()).round()

df
```

```python
df
```

```python
df.Salary = df.Salary.fillna(df.Salary.median())

df
```

```python
df.Name = df.Name.fillna(df['Name'].mode()[0])

df
```

```python
df.Score = df.Score.fillna(method='bfill')

df
```
