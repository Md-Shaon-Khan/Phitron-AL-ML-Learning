.loc[] হলো label-based indexing — row/column name দিয়ে select করে। .iloc[] হলো position-based — integer position দিয়ে select করে (0-based)। এই পার্থক্য না বুঝলে ML pipeline-এ ভুল data নিয়ে কাজ হবে।

| Concept               | Description                                                                                                                                                                  |
| --------------------- | ---------------------------------------------------------------------------------------------------------------------------------------------------------------------------- |
| **`.loc[]`**          | Label-based indexing। Row/column **name** ব্যবহার করে select করে। Slice-এর ক্ষেত্রে **inclusive**।                                                                           |
| **`.iloc[]`**         | Position-based indexing। Row/column **integer position (0-based)** ব্যবহার করে select করে। Slice-এর ক্ষেত্রে **exclusive end**।                                              |
| **`df['col']`**       | Single column selection। Returns a Series।                                                                                                                                   |
| **`df[['c1','c2']]`** | Multiple column selection। Returns a DataFrame।                                                                                                                              |
| **`.at[]`**           | Single value access **label-based**, fast।                                                                                                                                   |
| **`.iat[]`**          | Single value access **position-based**, fast।                                                                                                                                |
| **ML Application**    | - Features: `X = df.loc[:, feature_cols]`<br>- Labels: `y = df['target']`<br>- Train/test split: `df.iloc[train_idx]`<br>- Update single value: `df.at[row_label, col_name]` |

| Feature                | `loc`                                                            | `iloc`                                             |
| ---------------------- | ---------------------------------------------------------------- | -------------------------------------------------- |
| Indexing basis         | **Row/column labels** (name-based)                               | **Integer positions** (position-based)             |
| Row slice              | Includes **last label**                                          | Excludes **last index** (Python-style)             |
| Column slice           | Includes **last label**                                          | Excludes **last index**                            |
| Can use boolean arrays | ✅                                                                | ✅                                                  |
| Examples               | `df.loc[0:2, 'A':'C']` → rows with label 0,1,2 and columns A,B,C | `df.iloc[0:3, 0:3]` → rows 0,1,2 and columns 0,1,2 |
