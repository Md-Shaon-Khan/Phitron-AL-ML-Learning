df.dropna() — হ্যাঁ, পুরো row বাদ দেয় যদি সেই row-এর যেকোনো একটা cell-এও NaN থাকে।

df.dropna(subset=['Score', 'Age']) — পুরো column বাদ দেয় না। শুধু ওই দুটো column চেক করে। যদি কোনো row-এর Score বা Age cell-এ NaN থাকে, তাহলে সেই rowটা বাদ যায়। বাকি columns-এ NaN থাকলেও সমস্যা নেই।

df.dropna(thresh=2) — হ্যাঁ, ঠিকই ধরেছো। কোনো row-এ যদি কমপক্ষে ২টা valid (non-NaN) value থাকে, তাহলে সেই row রাখে। ২টার কম থাকলে বাদ দেয়।
