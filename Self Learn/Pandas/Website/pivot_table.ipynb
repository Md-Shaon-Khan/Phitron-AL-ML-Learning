```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'Month': ['Jan','Jan','Feb','Feb','Mar','Mar'],
    'Dept':  ['IT','HR','IT','HR','IT','HR'],
    'Sales': [100,80,120,90,150,110],
    'Profit':[20,15,25,18,30,22]
})
```

```python
df
```

```python
pt = df.pivot_table(
    values  = 'Sales',  
    index   = 'Month',    
    columns = 'Dept',     
    aggfunc = 'sum',    
    margins = True    
)
```

```python
pt
```
