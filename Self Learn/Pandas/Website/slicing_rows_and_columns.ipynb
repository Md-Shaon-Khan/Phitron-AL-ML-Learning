| Concept                    | Description                                                                                                           |
| -------------------------- | --------------------------------------------------------------------------------------------------------------------- |
| **Slicing**                | DataFrame-এর নির্দিষ্ট অংশ select করা। Row, column, বা উভয়ই slice করা যায়। Efficient slicing memory ও time বাঁচায়। |
| **Row slice**              | `df[start:end]` → rows `start` থেকে `end-1` select করে।                                                               |
| **Column & row slice**     | `df.loc[row_labels, col_labels]` → label-based selection।<br>`df.iloc[row_pos, col_pos]` → position-based selection।  |
| **SQL-like filter**        | `df.query("condition")` → condition অনুযায়ী rows filter করে।                                                         |
| **Head & Tail**            | `df.head(n)` → first n rows।<br>`df.tail(n)` → last n rows।                                                           |
| **Random sample**          | `df.sample(n, random_state=seed)` → random n rows select করে reproducibly।                                            |
| **Feature & target split** | `X = df.iloc[:, :-1]` → সব rows, last column বাদ।<br>`y = df.iloc[:, -1]` → last column only।                         |
