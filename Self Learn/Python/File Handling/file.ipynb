| Mode   | Name         | Description                                   | File Requirement        |
| ------ | ------------ | --------------------------------------------- | ----------------------- |
| `"r"`  | Read         | Opens file for reading only                   | File must exist         |
| `"w"`  | Write        | Opens file for writing (overwrites if exists) | Creates if not exists   |
| `"a"`  | Append       | Adds data at the end of file                  | Creates if not exists   |
| `"x"`  | Create       | Creates a new file                            | Fails if file exists    |
| `"r+"` | Read + Write | Read and write both                           | File must exist         |
| `"rb"` | Read Binary  | Read binary files (image, PDF, etc.)          | File must exist         |
| `"wb"` | Write Binary | Write binary files                            | Creates/overwrites file |

```python
lines = ['Line 1\n','Line 2\n','Line 3\n']

file = open('notes2.txt','w')
file.writelines(lines)
file.close()
```

| Feature      | `write()`                            | `writelines()`                  |
| ------------ | ------------------------------------ | ------------------------------- |
| Input Type   | Single string                        | List (or iterable) of strings   |
| Usage        | Writes one string at a time          | Writes multiple strings at once |
| Newline      | Must add `\n` manually               | Must also add `\n` manually     |
| Return Value | Returns number of characters written | Returns `None`                  |
| Flexibility  | More control (write step-by-step)    | Faster for multiple lines       |


```python
file = open('notes.txt','r')
content = file.read()
file.close()

print(content)
```

```python
file = open('notes2.txt','r')

line1 = file.readline()
line2 = file.readline()

print(line1.strip())
print(line2.strip())

file.close()
```

| Feature      | `read()`                          | `readline()`             |
| ------------ | --------------------------------- | ------------------------ |
| Purpose      | Reads entire file (or given size) | Reads one line at a time |
| Return Type  | String                            | String                   |
| Newline      | Includes `\n`                     | Includes `\n`            |
| Argument     | Optional (number of characters)   | No argument needed       |
| Usage        | `file.read()`                     | `file.readline()`        |
| Pointer Move | Moves to end (if full read)       | Moves to next line       |


```python
file = open('notes.txt','r')

lines = file.readlines()
file.close()

print(lines)
for line in lines:
  print(line.strip())
```

```python
file = open('notes.txt','r')

for line in file:
  print(line.strip())

file.close()
```

```python
with open('notes3.txt','w') as file1:
  file1.write("Hello from with block.\n")
  file1.write("File closes automatically.\n")
  
with open('notes3.txt','r') as file2:
  content = file2.read()
  print(content)
```

```python
with open('notes3.txt','a') as file:
  file.write('One more line please.\n')
```

```python
with open("students.csv", "w") as file:
    file.write("Name,Age,Grade\n")
    file.write("Alice,22,A\n")
    file.write("Bob,20,B\n")
    file.write("Charlie,23,A\n")
```

```python
with open("students.csv", "r") as file:
    lines = file.readlines()

header = lines[0].strip().split(",")
print(header) 

for line in lines[1:]:
    name, age, grade = line.strip().split(",")
    print(f"{name}  {age}  {grade}")
```

```python
import csv

with open('students.csv','w',newline="") as file:
  writer = csv.writer(file)
  writer.writerow(['Name','Age','Grade'])
  writer.writerow(['Shaon','22','A+'])
  writer.writerow(['Mahi',23,'B'])
  writer.writerow(['Nahid',23,'A'])
```

```python
with open("students.csv", "r") as file:
    reader = csv.reader(file)
    header = next(reader)        
    print("Header:", header)

    for row in reader:
        print(row)  
```

```python
with open("students.csv", "r") as file:
    reader = csv.DictReader(file)

    for row in reader:
        print(f"Name: {row['Name']} | Grade: {row['Grade']}")
```

| Function / Method             | Purpose / Description                               |
| ----------------------------- | --------------------------------------------------- |
| `open(filename, mode)`        | Open a file in specified mode (`r`, `w`, `a`, etc.) |
| `read(size)`                  | Read entire file or specified number of characters  |
| `readline()`                  | Read one line at a time                             |
| `readlines()`                 | Read all lines into a list                          |
| `write(string)`               | Write a single string to a file                     |
| `writelines(list_of_strings)` | Write a list (or iterable) of strings               |
| `close()`                     | Close the file (important to release resources)     |
| `with open(...) as f:`        | Context manager → automatically closes file         |
| `seek(offset)`                | Move file pointer to a specific location            |
| `tell()`                      | Returns current position of file pointer            |
| `flush()`                     | Flush buffer to file                                |
| `csv.reader(file)`            | Read CSV file rows as lists                         |
| `csv.DictReader(file)`        | Read CSV file rows as dictionaries                  |
| `csv.writer(file)`            | Write rows to CSV file                              |
| `csv.DictWriter(file)`        | Write rows as dictionaries to CSV file              |
| `os.remove(filename)`         | Delete a file                                       |
| `os.rename(src, dst)`         | Rename a file                                       |
| `os.path.exists(filename)`    | Check if file exists                                |
| `pickle.dump(obj, file)`      | Serialize Python object to a file                   |
| `pickle.load(file)`           | Deserialize Python object from file                 |
| `json.dump(obj, file)`        | Write Python object as JSON                         |
| `json.load(file)`             | Read JSON and convert to Python object              |
