```python
squares = []

for i in range(1,6):
  squares.append(i ** 2)
  
print(squares)

squares2 = [i ** 2 for i in range(1,6)]
print(squares2)
```

```python
nums = [1,2,3,4,5]

doubled = [n * 2 for n in nums]
print(doubled)

lowerLetter = ['a','b','c','d']
upper = [lowerLetter[c].upper() for c in range (len(lowerLetter))]
print(upper)
```

```python
nums = [2,1,4,5,3,5,6,8,9,4]

even = [num for num in nums if num % 2 == 0]
odd = [num for num in nums if num % 2 == 1]
big = [num for num in nums if num > 5]

print(even)
print(odd)
print(big)
```

```python
nums = [1,2,3,4,5,6,7,7,8,9,6,5,6,8]

levels = [(f"{num} - Even") if num % 2 == 0 else (f"{num} - Odd") for num in nums]
print(levels)
```

```python
sentence = "Hello Mr. Shaon Khan Boss"

chars = [c for c in sentence]
print(chars)

chars_without_space = [c for c in chars if c != ' ']
print(chars_without_space)

vowels = [c for c in sentence if c.lower() in "aeiou"]
print(vowels)
```

```python
pairs = []
for x in [1, 2, 3]:
    for y in ["a", "b"]:
        pairs.append((x, y))

# same thing with comprehension
pairs = [(x, y) for x in [1, 2, 3]   # outer loop first
                for y in ["a", "b"]]  # inner loop second

print(pairs)
```

```python
squares = {n: n**2 for n in range(1, 6)}   
print(squares)

words = ["apple", "banana", "kiwi"]
lengths = {w: len(w) for w in words}   
print(lengths)
```

```python
nums = [1, 2, 3, 4, 5, 6]

even_squares = {n: n**2 for n in nums if n % 2 == 0}
print(even_squares)
```

```python
original = {"name": "Alice", "city": "Dhaka", "job": "Dev"}

flipped = {v: k for k, v in original.items()}
print(flipped)
```

```python
nums = [1, 2, 2, 3, 3, 3, 4]

unique_squares = {n**2 for n in nums}  
print(unique_squares)

words = ["apple", "avocado", "banana", "blueberry", "cherry"]
first_letters = {w[0] for w in words} 
print(first_letters)
```
