```python
def greet():
  print("Hello World!")

greet()
```

```python
def show_line():
  print("_" * 30)

show_line()
show_line()
show_line()
```

```python
def greet(name):
  print(f"Hello, {name}")

greet("Shaon Khan")
greet("Ayan Khan")
```

```python
def add(a, b):
  print(a + b)

add(5,2)
add(10,20)
```

```python
def student_info(name, age, grade):
  print(f"{name} | Age:{age} | Grade:{grade}")
  
student_info(age=20,grade='A',name='Alice')
```

```python
def greet(name, message = "Good Morning"):
  print(f"{message}, {name}!")
  
greet('Shaon')
greet('Shaon Khan','Good Evening')
```

```python
def power(base,exp = 2):
  print(base ** exp)
  
power(3)
power(3,3)
```

```python
def total(*nums):
  print(sum(nums))

total(1,2,3)
total(10,20,30,40)
```

```python
def show_info(**details):
  for key,value in details.items():
    print(f"{key}: {value}")
  
show_info(name='Shaon',age=22,city='Dhaka')
```

```python
def add2(a, b):
  return a + b

result = add2(3, 5)
print(result)
```

```python
def min_max(nums):
  return min(nums), max(nums)

low, high = min_max([4,1,5,5,8])
print(low)
print(high)
```

```python
def check_age(age):
  if age < 0:
    return "Invalid age"
  if age < 18:
    return "Minor"
  return "Adult"


print(check_age(-1))
print(check_age(14))
print(check_age(76))
```

```python
count = 0
def increment():
  global count
  count += 1

increment()
increment()
increment()
print(count)
```

| Concept           | Example          | Description                                     |
| ----------------- | ---------------- | ----------------------------------------------- |
| Define Function   | `def func():`    | Creates a function                              |
| Call Function     | `func()`         | Executes the function                           |
| Parameter         | `def f(x):`      | Input passed to function                        |
| Default Parameter | `def f(x=5):`    | Uses default value if no argument is given      |
| *args             | `def f(*nums):`  | Accepts multiple positional arguments (tuple)   |
| **kwargs          | `def f(**info):` | Accepts multiple keyword arguments (dictionary) |
| Return            | `return value`   | Sends value back from function                  |
| Local Variable    | Inside function  | Accessible only within the function             |
| Global Variable   | Outside function | Accessible throughout the program               |
| global keyword    | `global x`       | Modify global variable inside function          |
