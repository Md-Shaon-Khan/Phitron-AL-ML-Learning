```python
info = {
  'x':1,
  'y':2,
  'z':3
}

print(info.keys())
print(info.values())
print(info.items())


for key,value in info.items():
  print(f"{key}: {value}")
```

```python
A = {1,2,3,4}
B = {3,4,5,6}

print(A | B)
print(A & B)
print(A - B)
print(A ^ B)
```

| Feature    | List              | Tuple               | Dictionary                           | Set             |
| ---------- | ----------------- | ------------------- | ------------------------------------ | --------------- |
| Syntax     | `[ ]`             | `( )`               | `{key: value}`                       | `{ }`           |
| Ordered    | Yes               | Yes                 | Yes (Python 3.7+)                    | No              |
| Mutable    | Yes               | No                  | Yes                                  | Yes             |
| Duplicates | Allowed           | Allowed             | Keys: Not allowed<br>Values: Allowed | Not allowed     |
| Indexing   | Yes               | Yes                 | By keys                              | No              |
| Use Case   | Sequence of items | Fixed/constant data | Key-value mapping                    | Unique elements |


| Feature / Operation | List                            | Tuple                  | Dictionary                 | Set                                    |
| ------------------- | ------------------------------- | ---------------------- | -------------------------- | -------------------------------------- |
| Create              | `[1, 2, 3]`                     | `(1, 2, 3)`            | `{"a":1, "b":2}`           | `{1, 2, 3}`                            |
| Insert/Add          | `append(x)`<br>`insert(i, x)`   | Not allowed            | `d[key] = value`           | `add(x)`                               |
| Delete              | `remove(x)`<br>`pop()`<br>`del` | Not allowed            | `pop(key)`<br>`del d[key]` | `remove(x)`<br>`discard(x)`<br>`pop()` |
| Update              | `list[i] = x`                   | Not allowed            | `update({...})`            | `update({...})`                        |
| Access              | Index (`list[i]`)               | Index (`tuple[i]`)     | Key (`d[key]`)             | Not supported                          |
| Search              | `x in list`                     | `x in tuple`           | `key in dict`              | `x in set`                             |
| Length              | `len(list)`                     | `len(tuple)`           | `len(dict)`                | `len(set)`                             |
| Clear all           | `clear()`                       | Not allowed            | `clear()`                  | `clear()`                              |
| Copy                | `copy()`                        | Not needed (immutable) | `copy()`                   | `copy()`                               |
| Special Notes       | Ordered, mutable                | Ordered, immutable     | Key-value pairs            | Unique, unordered                      |
