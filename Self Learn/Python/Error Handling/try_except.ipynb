```python
try:
  num = int(input("Enter a number: "))
  print(f"Entered number is: {num}")
except:
  print("That was not a valid number.")
```

```python
try:
  num = int('abc')
except ValueError as e:
  print(f'Invalid number format: {e}')
```

```python
try:
  my_list = [1,2,3,4]
  print(my_list[10])
except IndexError as r:
  print(f'Index out of range: {r}')
```

```python
try:
    num = int("abc")        
except ValueError:
    print("Error occurred!")
finally:
    print("This always runs!")
```

| Exception           | Cause                        | Example                  |
| ------------------- | ---------------------------- | ------------------------ |
| `ValueError`        | Wrong value type             | `int("abc")`             |
| `TypeError`         | Wrong data type              | `"a" + 1`                |
| `ZeroDivisionError` | Divide by zero               | `10/0`                   |
| `IndexError`        | List index out of range      | `[1,2,3][5]`             |
| `KeyError`          | Dictionary key doesn't exist | `d["x"]`                 |
| `FileNotFoundError` | File doesn't exist           | `open("file.txt")`       |
| `NameError`         | Variable not defined         | `x`                      |
| `AttributeError`    | Wrong method/attribute       | `"abc".push()`           |
| `ImportError`       | Module not found             | `import unknown_module`  |
| `Exception`         | Catch ALL exceptions         | `except Exception as e:` |
