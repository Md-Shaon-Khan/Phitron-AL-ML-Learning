```python
from datetime import datetime, date, timedelta

now = datetime.now()

print(now)
print(now.year)
print(now.month)
print(now.day)
print(now.strftime('%d-%m-%Y'))
print(now.strftime('%H:%M:%S %p'))
```

```python
today = date.today()

print(today)
```

```python
tomorrow = today + timedelta(days=1)
last_week = today - timedelta(days=7)

print(tomorrow)
print(last_week)
```

```python
birthday = date(2004,11,12)
today = date.today()

diff = today - birthday
print(diff.days)
```

| Syntax                        | Use / Purpose                          | Example Usage   |
| ----------------------------- | -------------------------------------- | --------------- |
| `import math`                 | Import whole module                    | `math.sqrt(16)` |
| `from math import sqrt`       | Import specific function/item          | `sqrt(16)`      |
| `from math import *`          | Import all functions (not recommended) | `sqrt(16)`      |
| `import math as m`            | Import module with alias               | `m.sqrt(16)`    |
| `from math import sqrt as sq` | Import specific item with alias        | `sq(16)`        |
